## Summary <a class="tocSkip"></a>
In this lecture we discuss Newton's method in multiple dimensions for solving nonlinear root-finding and optimisation problems. By using derivative information, Newton's method is more efficient than methods that do not use this information (e.g. Conjugate Gradient) and shows quadratic local convergence. Local convergence means that this is only seen near the solution. Global convergence, i.e. convergence starting from any point within the domain, can be obtained by combining Newton's method with line search or trust region methods. Trust region methods, although generally more complex than line search methods, are often more robust, in particular when dealing with non-convex problems. Computing the required derivative information, the Jacobian in root-finding problems or the Hessian for optimisation, can be expensive for large problems. For this reason we discuss some quasi-Newton methods in which the Jacobian/Hessian does not need to be calculated exactly, but is approximated. In this category, Gauss-Newton methods are used to solve nonlinear least-squares problems. More general problems can be solved using quasi-Newton secant methods such as Broyden's method and BFGS which will be discussed in lecture 9.

### Important concepts: <a class="tocSkip"></a>
- one-dimensional methods: Chord method, Newton's method and the secant method
- the role of the Jacobian, resp. Hessian in non-linear root-finding and optimisation problems
- line search methods that ensure sufficient decrease of the function value, and prevent step sizes that are too small
- the role of scaling of input parameters and output values, and the importance of scale invariance of the methods
- trust region methods which can be more robust than line search methods
- Newton Gauss (damped or Levenberg-Marquard) for nonlinear least-square problems, in particular nonlinear regression

# One-dimensional methods
In this section we will introduce some standard methods to solve $F(x)=0$ for nonlinear functions $F:\mathbb{R}\to\mathbb{R}$. In the next we will formalize these for the more general case $F(\boldsymbol{x})=\boldsymbol{0}$ for multi-dimensional functions $F:\mathbb{R}^n\to \mathbb{R}^n$.

We start with the simple idea of finding a fixed point $G(x)=x$ a function $G$.
A <a class="definition" href="#definitions" id="Picarditeration">fixed point iteration</a> or <a class="definition" href="#definitions" id="Picarditeration">Picard iteration</a> is an iterative method to find such a fixed point, by iteratively applying:

$$
  x^{(i+1)} = G(x^{(i)})
$$

that is we start with an initial guess $x^{(0)}$, then the next iterate is $x^{(1)}=G(x^{(0)})$, then

$$x^{(2)}=G(x^{(1)})=G(G(x^{(0)})),$$,
$$x^{(3)}=G(x^{(2)})=G(G(G(x^{(0)}))),$$
$$\ldots$$

and so on.

If we define a second function $F(x)=G(x)-x$, then we can see that finding a fixed point $x$ of $G$: $G(x)=x$, is equivalent to finding a point $x$ where $F(x)=0$. In other words the fixed point of $G$ is a _root_ of $F$. Reversely we can start with a root-finding problem $F(x)=0$ for arbitrary function $F$, and turn it into a fixed point problem $G(x)=x$ by defining $G(x)=x + F(x)$.

The same iteration in terms of the function $F$ looks like
$$
  x^{(i+1)} = x^{(i)} + F(x^{(i)})
$$

Of course for a cheap-to-evaluate function in one dimension we could try to solve the fixed point graphically by finding the intersection of the graphs $y=G(x)$ and $y=x$. Below we do just that for the function $G(x)=\exp(-x)$:

```python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)

x = np.linspace(0, 1, 100)

ax1.plot(x, np.exp(-x), 'b', label=r'$y = G(x) = \mathrm{e}^{-x}$')
ax1.plot(x, x, 'k--', label='$y=x$')

# solution obtained from Picard iteration below: 0.5671221400953808
ax1.plot(0.5671221400953808, 0.5671221400953808, 'ro', label='intersection')

ax1.legend(loc='best', fontsize=14)
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$y$', fontsize=16)
ax1.set_title('Fixed point as the intersection of two curves', fontsize=16);
```

Note that this requires the evaluation of $G(x)$ in many points, and to acquire an accurate answer we would problem need to zoom in and evaluate at many more points closer to the fixed point. Here, we already had a pretty good knowledge of the $y=\exp(-x)$ graph and roughly where it would intersect with $y=x$ line. For more general non-linear functions you may encounter in practice however, it is a lot harder to decide where to put the resolution, i.e. in which points to evaluate the function in order to get an idea where the intersection may occur. Taking into account the fact the function may be computationally expensive to evaluate, and that for multi-dimensional problems you would need many more point to "graph" the function, such approaches quickly become completely infeasible.

Let's see how well our simple Picard iteration fares. We start from an initial guess of $x^{(0)}=1$

```python
def picard_method(G, x_0, atol = 1.e-5, maxiter=100):
    x_n = []
    y_n = []
    x = x_0
    
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    # since we include the initial guess, the max. number of entries is maxiter+1
    for i in range(maxiter+1):
        x_n.append(x)
        y_n.append(G(x))
        # Picard iteration:
        x = G(x)
        if abs(x - x_n[-1]) < atol:
            break
    
    return x_n, y_n
```

```python
def G(x):
    return np.exp(-x)

x_guess = -1
x_n, y_n = picard_method(G, x_guess, atol=1.0e-5, maxiter=20)
fig, ax = plt.subplots(figsize=(8, 8))
plot_picard_iteration(ax, x_n, y_n, G, max_labels=3, flabel=r'$y$');

# print iterations:
print(f'x^(0) = {x_guess}')
prev_x = x_guess
for i, x in enumerate(x_n[1:]):
    print(f'x^({i+1}) = {x}, \t |x^({i+1})-x^({i+0})| = {abs(x-prev_x)}')
    prev_x = x
    
print(f'Solution: x = {x_n[-1]}, G(x) = {G(x_n[-1])}')
```

The method doesn't always converge however. As can be seen by comparing the following two cases:

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

def G(x):
    return np.sqrt(x)

x_guess = 0.4
x_n, y_n = picard_method(G, x_guess, atol=1.0e-2, maxiter=6)
plot_picard_iteration(ax[0], x_n, y_n, G, max_labels=3, flabel=r'$y$')

def G(x):
    return np.tan(x) - .1

# start very close to root using value found below - need to set atol lower (i.e. strict) otherwise stops at first it.
x_guess = 0.5
x_n, y_n = picard_method(G, x_guess, atol=1.0e-6, maxiter=12)
plot_picard_iteration(ax[1], x_n, y_n, G, max_labels=12, flabel=r'$y$')
```

Note that in the right-hand side case, $G(X)=\tan(x)-.1$, the initial guess is already very close to the solution, yet the Picard iteration gets further and further away from the solution (we say it _diverges_) instead of converging to the solution. This appears to be because the graph $y=G(x)$ "bends" away from the $y=x$ line.

### Fixed Point Theorem (\*) <a class="tocSkip"></a>
<div class="optional">
At $i$-th iteration, we can measure the vertical distance between the point at the graph at $(x^{(i)}, G(x^{(i)})$ and the point at the $y=x$-line $(x^{(i)}, x^{(i)})$, which are the vertical green dashed distances in the figures above

$$| G(x^{(i)}) - x^{(i)} |$$

However since $G(x^{(i)})$ is also the $x$ value for the next iterate $x^{(i+1)}$:

$$| G(x^{(i)}) - x^{(i)} | = | x^{(i+1)} - x^{(i)} | $$

this distance is equal to horizontal distance between $x^{(i)}$ and $x^{(i+1)}$. Check for yourself in the figures above that this indeed the case for all vertical and subsequent horizontal distances.

Of course, we can also say $x^{(i)}=G(x^{(i-1)})$ and thus we can describe the same distance as:

$$| G(x^{(i)}) - G(x^{(i-1)}) | = | x^{(i+1)} - x^{(i)} | $$

For the iteration to converge, we want it to be shorter than the previous distance between $x^{(i)}$ and $x^{(i-1)}$:

$$| G(x^{(i)}) - G(x^{(i-1)}) | = | x^{(i+1)} - x^{(i)} | \lt | x^{(i)} - x^{(i-1)} |$$

A function $G(x)$ for which

$$| G(x) - G(y) | \leq q |x- y|$$

for arbitrary $x, y$ in the domain of $G$, with a fixed constant $q, 0\leq q \lt 1$, is called a <a class="definition">contraction mapping</a>. Because $q$ is strictly smaller than 1, this is indeed sufficient for the subsequent iterations to get closer and closer:

$$| x^{(i+1)} - x^{(i)} | = | G(x^{(i)}) - G(x^{(i-1)}) | \leq q | x^{(i)} - x^{(i-1)} |$$

Not only the subsequent iterations are get closer, also the residual $x-G(x)$ is getting smaller and smaller, since equivalently:

$$| G(x^{(i)}) - x^{(i)} | \leq q | G(x^{(i-1)}) - x^{(i-1)} |$$

This means the iteration converges to a point where $G(x)=x$. This result is known as the <a class="definition">Banach Fixed Point Theorem</a> (also known as  contraction mapping theorem). This kind of convergence where a difference or residual is guaranteed to be reduced by a certain factor, is an example of linear convergence.

Note that for such a contraction mapping (assuming it's differentiable):

$$
| G'(x) | = \left| \lim_{h\to 0} \frac{G(x+h) - G(x)}{h} \right|
= \lim_{h\to 0} \frac{|G(x+h) - G(x)|}{h} \leq q\lim_{h\to 0} \frac{|x+h - x|}h = q
$$
</div>

In conclusion: the absolute value of the derivative of $G(x)$ should be smaller than 1 for it to be a contraction which guarantees convergence of the Picard iteration.

Returning to the root-finding problem $F(x) = 0$, we can modify the associated function $G(x)=x - F(x)$ to $G(x)=x - m F(x)$ with arbitrary nonzero constant $m$. For any choice of $m$ the fixed point problem $G(x) = x$ is still equivalent to the root-finding problem $F(x)=0$.

This modifies the iteration in terms of $F$ to:

$$
  x^{(i+1)} = x^{(i)} - m F(x^{(i)})
$$

This method is sometimes known as the <a class="definition" href="#definitions" id="Chordmethod">Chord method</a>.

The purpose of introducing the constant $m$, is that we can now better control the derivative of $G$:

$$
G'(x) = 1 - m F'(x)
$$

If we now choose $m$, such that

$$
  0 \leq m F'(x) \leq 2
$$

then

$$
-1 \leq G'(x) \leq 1 \implies | G'(x) | \leq 1
$$

Let's try this Chord method on $F(x) = x - \tan(x) + .1$. For $m=1$ this leads to the same fixed point iteration with $G(x) = 1 - m F(x) = \tan(x) - .1$ that we previously saw did not converge. For the values of $m=-0.4$ and $m=-1$ we see however that the Chord iteration now does converge. It does appear however that the convergence rate strongly depends on our choice of $m$:

```python
def chord_method(F, m, x_0, atol = 1.e-5, maxiter=100):
    x_n = []
    y_n = []
    x = x_0
    
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    # since we include the initial guess, the max. number of entries is maxiter+1
    for i in range(maxiter+1):
        x_n.append(x)
        y_n.append(F(x))
        # Chord update:
        x = x - m * F(x)
        if abs(x - x_n[-1]) < atol:
            break
    
    return x_n, y_n       

# note that with m=1, this gives G(x) = np.tan(x) - .1 as before
def F(x):
    return x - np.tan(x) + .1


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
fig.subplots_adjust(wspace=0.3)

ax1.axis('equal')
x0 = 0.6
m = -2.5
x_n, y_n = chord_method(F, m, x0, atol=1.e-2, maxiter=4)
plot_iteration(ax1, x_n, y_n, F)
plot_triangle(ax1, x_n, y_n, m)
ax1.set_title("Chord method iteration", fontsize=20)
print('The slope marker contains the value 1/m = ',1./m)

m = -1.
ax2.axis('equal')
x_n, y_n = chord_method(F, m, x0, atol=1.e-3, maxiter=4)
plot_iteration(ax2, x_n, y_n, F)
plot_triangle(ax2, x_n, y_n, m)
ax2.set_title("Chord method iteration", fontsize=20)
print('The slope marker contains the value 1/m = ',1./m)
```

Note that unlike the previous figures, we are now plotting $F(x)$, not $G(x)$, and we are now searching for the intersection of the graph of $F(x)$ with the $y=0$ line. If we connect each point on the graph $(x^{(i)}, F(x^{(i)})$ to the next iteration on the $y=0$ line: $(x^{(i+1)}, 0)$, we appear to obtain diagonal lines all with the same slope. In fact that slope can be determined to be:

$$
  \frac{\Delta y}{\Delta x} = \frac{F(x^{(i)}) - 0}{x^{(i)} - x^{(i+1)}} = 
  \frac{F(x^{(i)})}{x^{(i)} - x^{(i)} + m F(x^{(i)})} = \frac{1}{m}
$$

From experimenting with different values of $m$ you can see that the fastest convergence is obtained, if the diagonal line follows the tangent line of the graph. This idea leads to Newton's method. Rather than choosing a fixed $m$, we may also choose an $m^{(i)}$ that changes per iteration:

$$
  x^{(i+1)} = x^{(i)} - m^{(i)} F(x^{(i)})
$$

and the choice $m^{(i)}=1/F'(x^{(i)})$ gives <a class="definition" href="#definitions" id="Newtonsmethod">Newton's method</a> (also known as the <a class="definition" href="#definitions" id="NewtonRaphsonmethod">Newton-Raphson method</a>):

$$
  x^{(i+1)} = x^{(i)} - \frac{F(x^{(i)})}{F'(x^{(i)})}
$$

For this choice it can be shown that the method converges quadratically, which means that for some constant $K$:

$$|x^{(i+1)} - x^\ast| < K |x^{(i)} - x^\ast|^2$$

as long as we start _close enough_ to the solution $x^\ast$.

## Root finding vs. Optimisation
As we saw in lecture 2b the problem of finding a minimum of a function $f(x)$ 
typically involves finding the stationary points $x$ where $f'(x) = 0$ which is in fact a root-find problem on the function $F(x)=f'(x)$.

If we express the Chord method in terms of the function $f(x)$,

$$x^{(i+1)} = x^{(i)} - m F(x^{(i)}) = x^{(i)} - m f'(x^{(i)}),$$
  
we see that in fact it looks exactly like the Steepest (or Gradient) Descent method we encountered before.

Notice that in the Chord method we sometimes choose a _negative_ $m$ (as we did in the example above) - based on the criterion

$$
  0 \leq m F'(x) \leq 2 \quad\rightarrow\quad 0 \leq m f''(x) \leq 2
$$

which shows that in the optimisation case we should choose $m$ negative if the second derivative $f''(x)$ is negative. This result makes sense if you recall that if the second derivative is positive, a stationary point implies a local minimum, and thus we should go in the opposite gradient direction (like we did in lecture 2) with positive $m$. If the second derivative is negative however, a stationary point corresponds to a local maximum. If we want to use the algorithm to find such a maximum, we should be going uphill instead, in the same direction as the gradient, which requires $m$ to be negative.

Finally, we can also cast Newton's method as an optimisation algorithm

$$
  x^{(i+1)} = x^{(i)} - \frac{F(x^{(i)})}{F'(x^{(i)})}
  = x^{(i)} - \frac{f'(x^{(i)})}{f''(x^{(i)})}
$$

## Secant method

The analytical derivative $F'(x)$ might not always be available in which case we have to (numerically) approximate it. This is typically based on the (finite difference) formula:

$$
  F'(x) = \frac{F(x+h)-F(x)}{h} + \mathcal{O}(h)
$$

Rather than using multiple (potentially costly) function evaluations per iterations, we can instead choose to reuse the function evaluations from previous iterations for this:

$$
  F'(x^{(i)}) \approx  \frac{F(x^{(i)}) - F(x^{(i-1)})}{x^{(i)}-x^{(i-1)}} =: \frac 1{m^{(i)}}
$$

which results in an algorithm called the <a class="definition" href="#definitions" id="secantmethod">secant method</a>:

$$
  x^{(i+1)} = x^{(i)} - \frac{\left(x^{(i)}-x^{(i-1)}\right)F(x^{(i)})}{F(x^{(i)}) - F(x^{(i-1)})}
$$

Approximating the derivative in general leads to a reduction in the order of convergence of the method. In the case of the secant method it can be shown that the order of convergence drops from $2$ (i.e. quadratic convergence) to the golden ratio $\tfrac 12(1+\sqrt 5)\approx 1.618$.

# Newton's method in multiple dimensions

To extend Newton's (Newton-Raphson) method to multiple dimensions, solving ${\bf F}({\bf x})={\bf 0}$ for a function ${\bf F}:\mathbb{R}^n\to\mathbb{R}^n$, we derive it again from first principles. 

We start by writing down the Taylor expanson of ${\bf F}$ around the current best guess ${\bf x}^{(i)}$:

\begin{align*}
  {\bf F}({\bf x}) &= {\bf F}({\bf x}^{(i)}) + \sum_j\frac{\partial {\bf F}({\bf x}^{(i)})}{\partial x^{(i)}_j} \left(x_j - x^{(i)}_j\right) + \mathcal{O}\left(\left({\bf x} - {\bf x}^{(i)}\right)^2\right) \\[5pt]
    &= {\bf F}({\bf x}^{(i)}) + {\bf F}'({\bf x}^{(i)}) \left({\bf x} - {\bf x}^{(i)}\right) + \mathcal{O}\left(\left({\bf x} - {\bf x}^{(i)}\right)^2\right)
\end{align*}

where ${\bf F}'$ is defined as the $n\times n$ gradient matrix:

$$
\newcommand\gf[2]{\frac{\partial F_{#1}({\bf
x})}{\partial x_{#2}}}
  {\bf F}'({\bf x}) = \begin{pmatrix}
  \gf 11 & \gf 12 & \dots & \gf 1n \\
  \gf 21 & \gf 22 & \dots & \gf 2n \\
  \vdots & \vdots & \ddots & \vdots \\
  \gf n1 & \gf n2 & \dots & \gf nn
  \end{pmatrix}.
$$

This matrix is also referred to as the <a class="definition" href="#definitions" id="Jacobian">Jacobian</a>. Note that this matrix is not symmetric, and in some literature the  Jacobian is defined as the transpose of the above. In our definition the matrix vector multiplication of ${\bf F}'$ with a direction vector ${\bf p}$ gives the gradient in the direction of ${\bf p}$ (cf. the [directional derivative](https://en.wikipedia.org/wiki/Directional_derivative)).

Dropping the higher order terms, we obtain a linearization of ${\bf F}$ around ${\bf x}^{(i)}$ given by

$$
  {\bf F}_{\text{linear}}({\bf x}) = {\bf F}({\bf x}^{(i)}) + {\bf F}'({\bf x}^{(i)}) \left({\bf x} - {\bf x}^{(i)}\right)
$$

We now choose our next iterate ${\bf x}^{(i+1)}$ such that it satisfies:

$$
  {\bf F}_{\text{linear}}({\bf x}^{(i+1)}) = 0
$$

which leads to the following linear system:

$$
  {\bf F}'({\bf x}^{(i)}) {\bf p} = -{\bf F}({\bf x}^{(i)})
  \;\;\; \text{ with } \;\;\; {\bf p} = {\bf x}^{(i+1)} - {\bf x}^{(i)}
$$

Thus the $n$-dimensional Newton's method consists of the following steps for each iteration:

* evaluate ${\bf F}({\bf x}^{(i)})$
* evaluate ${\bf F}'({\bf x}^{(i)})$
* solve ${\bf F}'({\bf x}^{(i)}) {\bf p} = -{\bf F}({\bf x}^{(i)})$
* update ${\bf x}^{(i+1)} = {\bf x}^{(i)} + {\bf p}$

The algorithm can also be used to find stationary points of a function $f:\mathbb{R}^n\to\mathbb{R}$; By setting ${\bf F}(x)=f'({\bf x})$ we can find a point ${\bf x}\in\mathbb{R}^n$ sucht that $f'({\bf x})=0$. Here the usual caveat applies that not every stationary point of $f$ is a local minimum of $f$ (see the necessary and sufficient conditions of lecture 2b).

Note that if ${\bf F}(x)=f'({\bf x})$, then the gradient (matrix) of ${\bf F}$ is the Hessian (second derivative) of $f$. For this particular case we also know that ${\bf F}'$ is symmetric:

$$
  {\bf F}'({\bf x}) = {\bf F}'({\bf x})^T = f''({\bf x})
$$

where $f''$ in the case of multi-dimensional independent variables denotes the Hessian matrix.

Depending on whether Newton's method is used for a root-finding problem of the form ${\bf F}({\bf x})=0$, or whether it's used in optimisation (minimisation of $f$), this matrix is therefore referred to either as the Jacobian or the Hessian matrix.

For general functions $f$, the Hessian will not always be SPD. In particular it might not even be invertible, in which case Newton's method immediately breaks down. As we will see later on, the fact that the Hessian is not SPD means that the function $f$ does not necessarily decrease in the chosen direction ${\bf p}$!

## Local Convergence for Newton's method
In one dimension the strong point of Newton's method is the following: assuming we're close enough to a root/stationary point the method will converge quadratically to it. The same holds for Newton's method in multiple dimensions.

<div class="optional">
For a more precise mathematical statement of this *local convergence* property, we have to make the following assumptions:

1. $F$ is differentiable and its derivative $F'$ is [<a class="definition" href="#definitions" id="Lipschitzcontinuous">Lipschitz continuous</a>](https://en.wikipedia.org/wiki/Lipschitz_continuity), which means that there is a positive constant $\gamma$ such that for any two points $\bf x$, and $\bf y$

$$
  \| F'({\bf x}) - F'({\bf y}) \| \leq \gamma \|{\bf x} - {\bf y} \|
$$

Note that this is a weaker condition than $F'$ being a _contraction mapping_. For Lipschitz continuous we may use any positive scalar $\gamma$ but for a contraction mapping we require require $\gamma\lt 1$.

2. $F({\bf x}^*) = 0$


3. $F'({\bf x}^*)$ is nonsingular</div>

<div class="optional">
In the context of optimisations where $F({\bf x})=f'({\bf x})$, assumption 1. is often phrased  as $f$ being *twice Lipschitz continously differentiable*:

$$
  \| f''({\bf x}) - f''({\bf y}) \| \leq \gamma \|{\bf x} - {\bf y} \|
$$

Assumption 2. implies that ${\bf x}^*$ is a stationary point and if in addition to 3. we assume that $F'({\bf x})=f''({\bf x})$ is SPD then we know (see lecture 2b) that $f$ has a local minimum at ${\bf x}^*$.
</div>

Under these assumptions it can be proven, both for general nonlinear inversions $F({\bf x})=0$  and optimisation problems, that for a starting point ${\bf x}^{(0)}$ _close enough_ to ${\bf x}^*$, Newton's method converges quadratically to ${\bf x}^*$:

$$
  \| {\bf x}^{(n+1)} - {\bf x}^* \| \leq K \| {\bf x}^{(n)} - {\bf x}^* \|^2
$$

Unfortunately it is not generally possible to tell _how close_ the starting point needs to be to ensure this quadratic convergence.

# Line Search Methods

The steepest descent method for the minimisation of a function $f$, introduced in lecture 2b, took the following form

$$
  {\bf x}^{(i+1)} = {\bf x}^{(i)} + \lambda {\bf p}^{(i)}
$$

where ${\bf p}^{(i)}=-f'({\bf x}^{(i)})$, i.e. a step direction opposite to the gradient of $f$ in ${\bf x}^{(i)}$. 

The conjugate gradient method was also shown to take this form with a different choice of ${\bf p}^{(i)}$ (obtained by projecting the gradient to be $A$-orthogonal to previous search directions). Finding a suitable $\lambda$ is equivalent to finding a suitable new iterate ${\bf x}^{(i)}$ along the line spanned by the search direction ${\bf p}^{(i)}$ and is called a <a class="definition" href="#definitions" id="linesearch">line search</a>.

In lecture 2b, we simplified the analysis by assuming that $f$ is quadratic (and we saw that this quadratic minimisation problem is equivalent to a linear inversion). For this quadratic case we can easily find an optimal $\lambda$ such that $f({\bf x}^{(i+1)})$ is minimal; $f$ restricted to the line spanned by ${\bf p}^{(i)}$ becomes a simple one-dimensional quadratic function (recall that we used the notation $\hat f$ for this function). Note however that this is not the case for general nonlinear functions $f$. There we would have to solve a general one-dimensional nonlinear minimisation problem, typically requiring many evaluations of $f$, and this would have to be repeated for every iteration of the full dimensional problem. For this reason, in practice we use line search methods that do not necessarily result in the smallest possible new value $f({\bf x}^{(i+1)})$, but a new value that is _sufficiently_ smaller than the previous value $f({\bf x}^{(i)})$.

We define a <a class="definition" href="#definitions" id="descentdirection">descent direction</a>, to be any direction ${\bf p}$ such that

$$
  \left.\frac d{dt}\right|_{t=0} f({\bf x}+t{\bf p}) = f'({\bf x})^T {\bf p} \lt 0,
$$

that is any direction in which $f$ decreases for infinitesimally small step size. Clearly this is satisfied for the method of *steepest* descent, where ${\bf p}=-f'({\bf x})$. We will however consider a number of other methods for which in each iteration a search direction ${\bf p}^{(i)}$ is used that is a descent direction.

If ${\bf p}^{(i)}$ is a descent direction, it is always possible to choose a $\lambda$ small enough such that

$$
  f({\bf x}^{(n+1)}) \lt f({\bf x}^{(n)}).
$$

This means that if $f$ is bounded from below, the series $f({\bf x}^{(0)}), f({\bf x}^{(1)}), \dots$ converges
to a value

$$
  \lim_{k\to\infty} f({\bf x}^{(k)}) = f^*
$$

It is important to realize however that this does not necessarily mean that the method converges to a local minimum of $f$, as we can observe in the following two examples of minimizing $f(x)=x^2$.

```python
def f(x):
    return x**2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))

x_n =[1+2**-i for i in range(10)]
y_n =[f(x) for x in x_n]
plot_iteration(ax1, x_n, y_n, f, include_chords=False, left_extra=1.5)
ax1.set_ylabel('$f(x)$')

x_n =[(1+2**-i)*(-1)**i for i in range(10)]
y_n =[f(x) for x in x_n]
plot_iteration(ax2, x_n, y_n, f, include_chords=False)
ax2.set_ylabel('$f(x)$');
```

In both cases above, $\lim_{i\to\infty} f(x^{(i)})=1$ whereas of course the desired (local and global) minimum of $f$ is in $x=0$ with $f(0)=0$. In the left-hand case, the reason the algorithm stalls is that the step size $x^{(i+1)}-x^{(i)}$ is halved in each iteration, and does so before being sufficiently close to $x=0$. 
In the right-hand case, the step size does not go to zero, but the reduction in function value, $f(x^{(i+1)})-f(x^{(i)})$ does. Note that in both cases, with the appropriate choice of $\lambda$ these updates could be formulated as a steepest descent method since the chosen steps are always in the direction of the gradient.

It thus becomes clear that the requirement $f({\bf x}^{(i+1)})\lt f({\bf x}^{(i)})$ is not _sufficient_ to ensure convergence to a local minimum. It is not immediately obvious what a <a class="definition" href="#definitions" id="sufficientdecreasecriterion">sufficient decrease criterion</a> would look like. 

For any fixed value $\Delta f_{\text{min}}$ for a minimum decrease $f({\bf x}^{(i+1)})-f({\bf x}^{(i)})\lt-\Delta f_{\text{min}}$ the criterion will start failing as we approach a local minimum, simply because $f$ is already close enough to the minimum value that there is not enough room for $f$ to decrease any further. 
Similarly, if we're moving through an area of the domain where $f$ is very flat, we might be making large steps in ${\bf x}$ towards the desired minimum, but the change in $f$ will still be small. 

A more useful criterion, called the <a class="definition" href="#definitions" id="Armijiorule">Armijio rule</a>, therefore relates the change in $f$ to the expected change based on a linearisation of $f$:

$$
   f({\bf x}^{(i)} + \lambda{\bf p}^{(i)}) - f({\bf x}^{(i)}) \lt \alpha \lambda f'({\bf x}^{(i)})^T {\bf p}^{(i)}
$$

where $\alpha$ is a small value (often chosen to be $10^{-4}$). This criterion is called the *Armijo rule*.

## Wolfe Conditions and a Gradient-Angle Condition (\*)

<div class="optional">
The Armijo rule ensures sufficient decrease when "overshooting" the local minimum (as in the right-hand figure of the example above), but does not really stop us from choosing arbitrarily small $\lambda$. In convergence proofs the Armijo rule is therefore often combined with a second condition that prevents stalling because of $\lambda$ being chosen too small. An example are the *Wolfe conditions*:

\begin{align}
f(x^{(i)} + \lambda{\bf p}^{(i)}) - f(x^{(i)}) &\leq c_1 \lambda f'(x^{(i)})^T {\bf p}^{(i)} \\[10pt]
f'(x^{(i)} + \lambda{\bf p}^{(i)})^T{\bf p}^{(i)} &\leq c_2 f'(x^{(i)})^T {\bf p}^{(i)}
\end{align}

The first Wolfe condition is the Armijo rule that requires sufficient decrease in $f$. The second Wolfe condition also demands there to be some decrease in the derivative of the new point. This prevents choosing arbitrarily small $\lambda$, as we can see in the following sketch: the blue line represent the function $f$ restricted to the search direction. The first Wolfe condition restricts to below the orange dashed line. To avoid excessively small steps, we demand the slope of $f$ in the search direction to have reduced to a fraction, $c_2$ of the original slope at $x^{(i)}$.
</div>

```python
# coefficients for Wolfe conditions:
c1 = 0.01
c2 = 0.3

def f(x):
    return (x-1.5)*x*(x-3)*(x-4) - 4*x
def dfdx(x):
    return 4.0*x**3 - 25.5*x**2 + 45.0*x - 22.0
# second Wolfe condition: positive means satisfies condition
def wolfe2(x):
    return dfdx(x) - c2*dfdx(0)

# use Newton's method to find first point after x=0 that satisfies Wolfe 2
xwf2 = sop.newton(wolfe2, 0)
print("Second Wolfe condition satisfied in:", xwf2)
print("In this point dfdx = {} and c2*dfdx(0) = {}".format(dfdx(xwf2), c2*dfdx(0)))

x = np.linspace(0, 5, 100)
plt.figure(figsize=(16,6))
plt.plot(x, f(x), label='$f(x^{(k)}+\lambda p^{(k)})-f(x^{(k)})$')
plt.plot([0, x[-1]], [0,0], 'k-')
plt.plot([x[0], x[-1]], [x[0]*dfdx(0), x[-1]*dfdx(0)], 'k--', label='tangent line at $\lambda=0$')
plt.plot([x[0], x[-1]], [x[0]*c1*dfdx(0), x[-1]*c1*dfdx(0)], '--', label='1st Wolfe cond. (Armijo rule)')
plt.plot([x[0], x[-1]], 
         [f(xwf2)+(x[0]-xwf2)*c2*dfdx(0), f(xwf2)+(x[-1]-xwf2)*c2*dfdx(0)], '--',
         label='2nd Wolfe condition')
plt.plot(xwf2, f(xwf2), 'g.')
plt.axis([0, 5, -18, 1])
plt.xlabel('$\lambda$')
plt.legend();
```

<div class="optional">
Although the Wolfe conditions are sufficient conditions for the convergence of the steepest descent method, there is still one more condition needed for more general methods that are based on a descent direction ${\bf p}^{(i)}$ in every iteration. Although $f'({\bf x}^{(i)})^T {\bf p}^{(i)} \lt 0$ implies that the angle $\theta^{(i)}$ between the negative gradient $-f'({\bf x}^{(i)})$ and ${\bf p}$ is bounded by:

$$
  \cos(\theta^{(i)}) = \frac{-f'({\bf x}^{(i)})^T {\bf p}^{(i)}}{\|f'({\bf x}^{(i)})^T\| \|{\bf p}^{(i)}\|} > 0 \implies -90^\circ\lt\theta^{(i)}\lt90^\circ,
$$

it is possible that

$$
  \lim_{i\to\infty} \cos(\theta^{(i)}) = 0,
$$

in other words: although ${\bf p}^{(i)}$ is guaranteed to always have at least a component in the negative gradient direction, it is possible that it is slowly turning away from the negative gradient and thus for the method to stall.

In the following artificial example, where we are minimizing the trivial function $f(x)=x_0^2+x_1^2$, we choose a search direction that is composed of the negative gradient and a perpendicular component that grows bigger and bigger as $\|{\bf x}\|\to 1$. Therefore all search directions are descent directions, but they are slowly turning away from the negative gradient to become near orthogonal to it. Thus the iteration never converges to the minimum at ${\bf x}^*=(0,0)$ but keeps circling around whilst getting closer to the unit circle.
</div>

```python
def linesearch_quad(f0, f1, fp0):
    # f = f0 + fp0*x + (f1-f0-fp0)*x**2
    # f' = fp0 + 2*(f1-f0-fp0)*x = 0
    return -fp0/(f1-f0-fp0)/2.

def silly_method(f, jac, x0, maxiter=100):
    x_n = []
    y_n = []
    x = x0
    for i in range(maxiter+1):
        x_n.append(x)
        y_n.append(f(x))
        g = -np.array(jac(x))
        p_i = g  + np.array([x[1], -x[0]]) / (f(x)-1)**2
        alpha = linesearch_quad(y_n[-1], f(x+p_i), np.dot(-g, p_i))
        x = x + alpha*p_i        
    return x_n, y_n

def f(x):
    return x[0]**2 + x[1]**2
def jac(x):
    return [2*x[0], 2*x[1]]
x_n, y_n = silly_method(f, jac, [0,1.2], maxiter=500)
x_n = np.array(x_n)
plt.plot(x_n[:,0], x_n[:,1], 'g.');        
```

<div class="optional">
If we combine the Wolfe conditions with a strict lower bound on the $\cos(\theta^{(i)})$ we can finally prove [reference] the following global convergence guarantee:

Let $f:\mathbb{R}^n\to\mathbb{R}$ be bounded from below, differentiable and have Lipschitz continuous derivative $f'$. Then any method

$$
  {\bf x}^{(i+1)} = {\bf x}^{(i)} + \lambda^{(i)} {\bf p}^{(i)}
$$

where ${\bf p}^{(i)}$ satisfies:

$$
  \cos(\theta^{(i)}) = \frac{-f'({\bf x}^{(i)})^T {\bf p}^{(i)}}{\|f'({\bf x}^{(i)})^T\| \|{\bf p}^{(i)}\|} > \delta
$$

for a fixed $\delta>0$, combined with a line search method for $\lambda$ that satisfies the Wolfe conditions, we have that

$$
  \lim_{i\to\infty} f'({\bf x}^{(i)}) = {\bf 0}
$$

Note that this still does not guarantee that ${\bf x}^{(i)}$ converges to a single point ${\bf x}^*$ in which $f'({\bf x})=0$ (stationary point), but we are guaranteed to get arbitrarily close to either a point or region with zero gradient $f'$. 

This result is referred to as *global convergence* since the convergence holds true for an initial guess ${\bf x}^{(0)}$ anywhere in the domain (it does not refer to finding a *global minimum*). To ensure convergence to a strict minimizer further restrictions on the Hessian $f''$ are required.

The Wolfe conditions combined with a lower bound for $\theta^{(i)}$ are frequently used in global convergence proofs for optimisation problems. Sometimes other assumptions are made however, in particular for general nonlinear inversion problems of the form $F({\bf x})={\bf 0}$ for which there is not necessarily a function $f$ such that $f'=F$. There is typically however some form of the Armijo rule applied in the line search. 

Instead of strictly applying an explicit rule, like the second Wolfe condition, to avoid too small steps, it is often sufficient to start with a large enough initial step ${\bf p}^{(i)}$ with $\lambda=1$ (this depends on specific properties of the method of constructing ${\bf p}^{(i)}$), and reduce $\lambda$ by a constant factor until the Armijo rule is satisfied. This is called <a class="definition" href="#definitions" id="backtracking">backtracking</a>.

Another often used line search method makes efficient use of the information about $f$ along the line $f$. If we have evaluated the initial step with $\lambda=1$ but failed the Armijo rule, we already know the values of $f({\bf x}^{(i)})$ and $f({\bf x}^{(i)}+{\bf p}^{(i)})$. We also know the gradient of $f$ at ${\bf x}^{(i)}$, in particular in the ${\bf p}^{(i)}$ direction. This is enough information to approximate $f$ along the search line with a quadratic function. We can use this quadratic approximation to make a fairly accurate estimate of the location of the minimum of $f$ along the line, which we use as our next attempt for $\lambda$. If after evaluating $f$ for this value of $\lambda$, it also fails the test, we have another data point and we can fit a cubic polynomial. This method of quadratic or cubic polynomial fitting can also be used in a *backtracking* manner: whenever the minimum of the current polynomial is rejected as a new $\lambda$, a new polynomial is fitted over a shorter interval. To avoid choosing too small $\lambda$s the minimum of the new polynomial is taken on the interval $\left[ \beta_{\text{min}} \lambda, \beta_{\text{max}} \lambda\right]$ where $\lambda$ is the last attempt and $0\lt\beta_{\text{min}}\lt\beta_{\text{max}}\lt0$. This *safe guarding* is needed to ensure global convergence.
</div>

## Line Search for Newton's method

There is no global convergence guarantee for the Newton-Raphson method. As we saw above, the method has very good *local convergence* properties (quadratic convergence), but we need an initial guess that is close enough to the root/stationary point. 

It is easy to see why Newton's method might not be convergent everywhere. If we examine the step vector ${\bf p}^{(i)}$

$$
  {\bf p}^{(i)} = -f''({\bf x}^{(i)})^{-1} f'({\bf x}^{(i)})
$$

and examine whether ${\bf p}^{(i)}$ is a descent direction:

$$
  f'({\bf x}^{(i)})^T {\bf p}^{(i)} = -f'({\bf x}^{(i)})^T f''({\bf x}^{(i)})^{(-1)} f'({\bf x}^{(i)}) \lt 0
$$

This can only be guaranteed if $f''({\bf x}^{(i)})^{-1}$ is Symmetric Positive Definite (SPD). This is equivalent to saying that $f''({\bf x}^{(i)})$ is SPD. Thus it is very possible (when $f''$ is not SPD) that Newton's method sends us in a direction in which $f$ increases rather than decreases.

Even when $f''$ is SPD everywhere, Newton's method is not guaranteed to converge. We can see this in a simple one-dimensional example where are trying to find the roots of $F(x)=\operatorname{atan}(x)$. We can also find a primitive function $f$:

$$
  f(x) = x \operatorname{atan}(x) - \tfrac 12\log(x^2 + 1)
$$

such that $f'(x)=F(x)$, so that we can also treat this as a minimisation problem for $f$. The derivative is given by

$$
  f''(x) = F'(x) = \frac 1{1+x^2},
$$

and it is easy to see that $f''(x)>0$ everywhere, which is the 1D equivalent of saying that $f''(x)$ is SPD. Nonetheless, in the example below we see that if our initial guess is too far away from the root at $x=0$, the method diverges.

```python
def newton_method(F, jac, x_0, atol = 1.e-5, maxiter=100):
    x_n = []
    y_n = []
    x = x_0
    
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    # since we include the initial guess, the max. number of entries is maxiter+1
    for i in range(maxiter+1):
        x_n.append(x)
        Fx = F(x)
        y_n.append(Fx)
        # Newton update:
        x = x - Fx/jac(x)
        if abs(x - x_n[-1]) < atol:
            break
    
    return x_n, y_n


def F(x):
    return np.arctan(x)


def dFdx(x):
    return 1/(1+x**2)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
x0 = 1.0
x_n, y_n = newton_method(F, dFdx, x0, atol=1.e-2, maxiter=4)
plot_iteration(ax1, x_n, y_n, F)
ax1.set_title("Newton method iteration", fontsize=20)

x0 = 2.0
x_n, y_n = newton_method(F, dFdx, x0, atol=1.e-2, maxiter=2)
plot_iteration(ax2, x_n, y_n, F)
ax2.set_title("Newton method iteration", fontsize=20);
```

```python
# plot the exact same Newton iteration along with the primitive of F
# i.e. df(x)/dx = F(x) - so that we treat it as a minimisation of f
def f(x):
    return x*np.arctan(x) - np.log(x**2 + 1)/2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
x0 = 1.0
x_n, y_n = newton_method(F, dFdx, x0, atol=1.e-2, maxiter=4)
f_n = [f(x) for x in x_n]
plot_iteration(ax1, x_n, f_n, f, include_chords=False)
ax1.set_ylabel('$y=f(x)$')
ax1.set_title("Newton method iteration", fontsize=20)

x0 = 2.0
x_n, y_n = newton_method(F, dFdx, x0, atol=1.e-2, maxiter=2)
f_n = [f(x) for x in x_n]
plot_iteration(ax2, x_n, f_n, f, include_chords=False)
ax2.set_ylabel('$y=f(x)$')
ax2.set_title("Newton method iteration", fontsize=20);
```

As it turns out, the line search method can be used to ensure convergence in areas where the Jacobian/Hessian is SPD. Thus we modify Newton's method to:


* evaluate ${\bf F}({\bf x}^{(i)})$


* evaluate ${\bf F}'({\bf x}^{(i)})$


* solve ${\bf F}'({\bf x}^{(i)}) {\bf p} = -{\bf F}({\bf x}^{(i)})$


* perform line search to find $\lambda^{(i)}$ with ${\bf x}^{(i+1)} = {\bf x}^{(i)} + \lambda^{(i)} {\bf p}$


It can be shown that if we are near enough to a root $F({\bf x}^*)=0$ and the Hessian/Jacobian is SPD, the first attempt in the line search with $\lambda=1$ will be successful. This means that as we are getting closer to the solution, at some point we are again applying the original Newton's method and thus we should expect the same local quadratic convergence. Thus we have improved Newton's method considerably by combining its local convergence properties with global convergence properties, as long as we know that $f''$ is SPD everywhere.

In the pure Newton method without a line search the only pieces of information we need to provide are ${\bf F}({\bf x})$ and the ${\bf F}'({\bf x})$. Thus when we use it as an optimisation algorithm to find the minimum of $f$, by solving for $f'({\bf x})={\bf F}({\bf x}) =0$, we only need to provide $f'({\bf x})$ and the Hessian $f''({\bf x})$. With a line search however we do need provide the function $f$ itself to verify Armijo's rule. For root-finding problems that cannot be written as an optimisation problem (i.e. where there is no primitive function $f({\bf x})$ such that $f'({\bf x})=F({\bf x})$), we modify Armijo's rule to:

\begin{align}
   \| F({\bf x}^{(i)} + \lambda {\bf p}^{(i)}) \|^2 - \| F({\bf x}^{(i)}) \|^2
   &\leq \left.\alpha\lambda \frac{d\| F({\bf x}^{(i)} + \lambda {\bf p}^{(i)}) \|^2}{d\lambda}\right|_{\lambda=0} \\
   &= 
   2\alpha\lambda  \left(F'({\bf x}^{(i)}) F({\bf x}^{(i)}\right)^T {\bf p}^{(i)} \\
   &= 2\alpha\lambda F({\bf x}^{(i)})^T F'({\bf x}^{(i)}) {\bf p}^{(i)} \\
   &= -2\alpha\lambda F({\bf x}^{(i)})^T F(x) \\
   &= -2\alpha\lambda \|F({\bf x}^{(i)})\|^2
\end{align}

in other words we demand sufficient reduction in the l2-norm as a percentage of the previous l2-norm of $F$. In this derivation we have used that we know that $F'({\bf x}^{(i)}) {\bf p}^{(i)} = -F({\bf x}^{(i)})$ for Newton's method.

Let's return to our example above where the original Newton method diverged for points that are not near the root $x=0$, and study the behaviour of Newton with line search where the line search simply halves lambda until the Armijo rule is satisfied. As expected, the method is now convergent for any starting point, and the reduction of $\lambda$ is only applied further away from the solution, thus maintaining quadratic convergence for points that are closer.

```python
def simple_line_search(x, p, f, initial_slope, alpha=1e-4, max_search_iter=100):
    """Simple line search that halves lambda until the Armijo rule is satisfied
    
    x - initial point (at lambda=0)
    p - search direction, search is along x+lambda*p
    f - (callback) function to minimize
    initial_slope - derivative of f in x in p-direction"""
    lamda = 1.
    if initial_slope==0.0:
        return x, 1.0, 0
    for j in range(max_search_iter):
        xt = x + lamda*p
        # check Armijo rule:
        if f(xt) - f(x) <= alpha*lamda*initial_slope:
            break
        # uncomment to follow convergence of line search:
        # print(j,lamda,f(xt)-f(x), alpha*lamda*initial_slope)
        lamda /= 2.
    else:
        raise Exception("Line search did not converge")
    return xt, lamda, j


def newton_method_ls(F, jac, f, line_search, x_0, atol = 1.e-5, maxiter=100, verbose=False):
    "Newton method with line search"
    x_n = []
    y_n = []
    x = x_0
    
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    # since we include the initial guess, the max. number of entries is maxiter+1
    for i in range(maxiter+1):
        x_n.append(x)
        Fx = F(x)
        y_n.append(Fx)
        # Newton update:
        if isinstance(Fx, float) or len(Fx)==1:
            p = -Fx/jac(x)
        else:
            p = sl.solve(jac(x), -Fx)
                
        x, lamda, reductions = line_search(x, p, f, np.dot(F(x), p))
        if verbose:
            print("In iteration {}, x={}; {} reductions were needed (lambda={})".format(i, x_n[-1], reductions, lamda))
        if sl.norm(x - x_n[-1]) < atol:
            break
    
    return x_n, y_n
```

```python
def f(x):
    return x*np.arctan(x) - np.log(x**2 + 1)/2


def F(x):
    return np.arctan(x)


def dFdx(x):
    return 1/(1+x**2)


fig, ax = plt.subplots(1, 2, figsize=(16,6))
x0 = 2.0
x_n, y_n = newton_method_ls(F, dFdx, f, simple_line_search, x0, atol=1.e-5, verbose=True)
plot_iteration(ax[0], x_n, y_n, F)
ax[0].set_title("Newton method iteration", fontsize=20)

print()
x0 = 10.0
x_n, y_n = newton_method_ls(F, dFdx, f, simple_line_search, x0, atol=1.e-5, verbose=True)
plot_iteration(ax[1], x_n[:6], y_n[:6], F)
ax[1].set_title("Newton method iteration", fontsize=20);
```

## Scale invariance

Let us now compare the performance of the steepest descent method with Newton's method. 

For optimisations of $f$ with steepest descent we have

$$
  {\bf x}^{(i+1)} = {\bf x}^{(i)} -\lambda^{(i)} F({\bf x}^{(i)})
  = {\bf x}^{(i)} -\lambda^{(i)} f'({\bf x}^{(i)})
$$

where we now choose $\lambda^{(i)}$ via a line search method. We should notice that this is just the Chord method, which we have encountered in 1D with a fixed $\lambda=1/m$, which can be applied to more general root-finding problems $F({\bf x})=0$ as well. Having a function $f$ to measure sufficient decrease in the line search however does generally improve the line search.

Before we return to our sample problem, we introduce one change, an overall scaling of the function by a factor $C$:

\begin{align*}
  f(x) &= C \left(x \operatorname{atan}(x) - \log(x^2 + 1)/2\right), \\
  F(x) &= f'(x) = C \operatorname{atan}(x), \\
  F'(x) &= f''(x) = \frac C{1+x^2}.
\end{align*}

```python
def steepest_descent_ls(F, f, line_search, x_0, atol = 1.e-5, maxiter=100, verbose=False):
    "Steepest decent method with line search"
    x_n = []
    y_n = []
    x = x_0
    
    # iterate until we hit break either as we hit tolerance or maximum number iterations
    # since we include the initial guess, the max. number of entries is maxiter+1
    for i in range(maxiter+1):
        x_n.append(x)
        Fx = F(x)
        y_n.append(Fx)
        # Chord update:
        p = -Fx       
        x, lamda, reductions = line_search(x, p, f, F(x)*p)
        if verbose:
            print("In iteration {}, x={}; {} reductions were needed (lambda={})".format(i, x_n[-1], reductions, lamda))
        if sl.norm(x - x_n[-1]) < atol:
            break
    
    return x_n, y_n
```

```python
# NOTE: the three functions below depend on C which is set outside
counter = 0  # counts the number of f-evaluations
def f(x):
    # to change a value outside a function scope, it needs to be declared global:
    global counter  
    counter += 1
    return C*(x*np.arctan(x) - np.log(x**2 + 1)/2)


def F(x):
    return C*np.arctan(x)


def dFdx(x):
    return C/(1+x**2)


fig, ax = plt.subplots(1, 1, figsize=(16,6))
for C in [0.01, 0.1, 0.5, 1., 10., 100.]:
    x0 = 10.0
    x_n, y_n = newton_method_ls(F, dFdx, f, simple_line_search, x0, atol=1.e-12)
    ax.semilogy(np.abs(x_n), '.', label='Newton C={}'.format(C))
    
    # we compute the number of function evaluations in the steepest descent method
    # by registering the counted n/o evaluations before and after
    counter_before = counter
    x_n, y_n = steepest_descent_ls(F, f, simple_line_search, x0, atol=1.e-12)
    counter_after = counter
    print("C = {}, number of function evaluations: {}".format(C, counter_after-counter_before))
    ax.semilogy(np.abs(x_n), '-', label='Chord    C={}'.format(C))
    ax.annotate('C={}'.format(C), [len(x_n), np.abs(x_n[-1])])
    
ax.set_xlabel('iterations')
ax.set_ylabel('error: $\|x\|$')
ax.set_title('Convergence of Newton and Chord methods under different scaling')
ax.legend();
```

We observe that the convergence of Newton's method is completely independent of the chosen scale $C$. This can be explained by writing out the effect of multiplying $f$, $f'$ and $f''$ by $C$ on the algorithm (check this for yourself!). This is an important property for optimisation and root-finding algorithms. Although it is good advice to rescale the input and output parameters (i.e. scaling of ${\bf x}$ and $F({\bf x})$) in such a way that values of order 1 are expected - for instance by choosing sensible units or [non-dimensionalisation](https://en.wikipedia.org/wiki/Nondimensionalization) - in many practical applications it is hard to control the extent of feasible values.

The same cannot be said for the Chord method (Steepest Descent). For small values of $C$ the method struggles to reduce $f$ within the 100 maximum allowed iterations. This is because, unlike for the Newton method, the step-size is directly proportional to $C$, resulting in small steps when $C$ is small. For large values of $C$ it seems to do better in terms of iterations but when looking at the number of function evaluations that are required, we see that this goes up rapidly, making the method still very costly. This is because for large $C$ the initial step size in the line search at $\lambda=1$ is far too large, which means we need a large number of reductions to satisfy the Armijo rule. Only at $C=1$ do we see convergence that is comparable (even slightly better) than Newton. This should be seen as a coincidence: for $C=1$, $F'(x)\to 1$ when $x\to 0$, which means that the Chord method and Newton's method convergence to the same step size!

# Trust Region Methods
Remember that Newton's method is defined using a linearisation of $F$ around the current guess ${\bf x}^{(i)}$:

$$
  F({\bf x}^{(i)} + {\bf p}) \approx F_{\text{linear}}({\bf x}^{(i)}+{\bf p}) = F({\bf x}^{(i)}) + F'({\bf x}^{(i)}) {\bf p}
$$

where ${\bf p}={\bf x}^{(i+1})-{\bf x}^{(i)}$ which we find by finding the root of this linearisation: $F_{\text{linear}}({\bf x}^{(i)}+{\bf p})={\bf 0}$. If $F({\bf x})=f'({\bf x})$ (optimisations), an equivalent description is to say that we approximate $f$ by a quadratic:

$$
  f({\bf x}^{(i)} + {\bf p}) \approx f_{\text{quadratic}}({\bf x}^{(i)}+{\bf p}) = f({\bf x}^{(i)}) + f'({\bf x}^{(i)})^T {\bf p} + \tfrac 12{\bf p}^T f''({\bf x}^{(i)}) {\bf p}
$$

and we find $p$ through minimisation of $f_{\text{quadratic}}$. If you compare the quadratic approximation with the stand quadratic form of lecture 2b you can see that this minimisation leads
to exactly the same Newton step.

If $f$ is indeed close to quadratic then we expect Newton's method to be very accurate without the need to decrease the step length (as we do in a line search). If the ${\bf p}$ obtained from Newton's original method however fails the test (Armijo rule) this indicates that $f$ is deviating too much from $f_{\text{quadratic}}$ and we should look for a new point closer to the previous one at ${\bf x}^{(i)}$. The trust region method, instead of simply shortening the proposed step vector
${\bf p}$ by a scalar factor, chooses a new step vector by minimizing the quadratic approximation in a smaller neighbourhood, the trust region. This may lead to a shorter step vector that doesn't necessarily still point in the same direction as the original proposed vector.

## Example and derivation (\*) <a class="tocSkip"><a/>
<div class="optional">
Let's look at an example minimisation in 2D of

$$
  f(x, y) = \frac{\left(x^2 + (\gamma y)^2 -1\right)^2}{x^2 + (\gamma y)^2 - 1/h} - y,
$$

where $\gamma$ determines the aspect ratio, and $h$ the height of the "bump" at ${\bf x}=(0,0)$. We study the behaviour of Newton's method with an intial guess at ${\bf x}_0=(-.5, -.5)$. First let's plot the contours and a 3D surface plot.
</div>

```python
from mpl_toolkits.mplot3d import Axes3D

gamma = 5  # aspect ratio
mx = 2.5  # width of axis view
x0 = np.array([-0.5,-0.5])  # initial guess
h = 10.  # height of bump

# here and below we ... (three dots) to leave free any unspecified dimensions
# so if x.shape=(D,):    x[...,0] = x[0]
#    if x.shape=(N,D):   x[...,0] = x[:,0]
#    if x.shape=(M,N,D): x[...,0] = x[:,:,0], etc.
# in this way the function f works regardless of whether we give it one point in R^D
# or a MxN raster of points
def f(x):
    return (x[...,0]**2 + (gamma*x[...,1])**2 - 1)**2/(x[...,0]**2 + (gamma*x[...,1])**2 + 1/h) - x[...,1]


def dfdx(xy):
    x = xy[...,0]
    y = xy[...,1]
    ans = np.empty_like(xy)
    ans[...,0] = -2*x*(gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h)**2 + 4*x*(gamma**2*y**2 + x**2 - 1)/(gamma**2*y**2 + x**2 + 1/h)
    ans[...,1] = -2*gamma**2*y*(gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h)**2 + 4*gamma**2*y*(gamma**2*y**2 + x**2 - 1)/(gamma**2*y**2 + x**2 + 1/h) - 1
    return ans


def hess(xy):
    x = xy[...,0]
    y = xy[...,1]
    ans = np.empty(xy.shape+(xy.shape[-1],))
    ans[...,0,0] = 2*(2*gamma**2*y**2 + 4*x**2*(gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h)**2 - 8*x**2*(gamma**2*y**2 + x**2 - 1)/(gamma**2*y**2 + x**2 + 1/h) + 6*x**2 - (gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h) - 2)/(gamma**2*y**2 + x**2 + 1/h)
    ans[...,0,1] = 8*gamma**2*x*y*((gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h)**2 - 2*(gamma**2*y**2 + x**2 - 1)/(gamma**2*y**2 + x**2 + 1/h) + 1)/(gamma**2*y**2 + x**2 + 1/h)
    ans[...,1,0] = ans[...,0,1]
    ans[...,1,1] = 2*gamma**2*(4*gamma**2*y**2*(gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h)**2 - 8*gamma**2*y**2*(gamma**2*y**2 + x**2 - 1)/(gamma**2*y**2 + x**2 + 1/h) + 6*gamma**2*y**2 + 2*x**2 - (gamma**2*y**2 + x**2 - 1)**2/(gamma**2*y**2 + x**2 + 1/h) - 2)/(gamma**2*y**2 + x**2 + 1/h)
    return ans
```

```python
x = np.linspace(-mx, mx, 40)
y = np.linspace(-mx/gamma, mx/gamma, 40)
# xy are the locations of the horizontal grid, it is a Nx x Ny x 2 array
xy = np.array(np.meshgrid(x, y)).T
fig = plt.figure(figsize=(12,4))
ax = fig.add_subplot(1, 1, 1)
cvalues = np.arange(-2,10,.5)
c = ax.contour(xy[...,0], xy[...,1], f(xy), cvalues)
ax.clabel(c, np.arange(0,10,1), fmt='%1.1f', colors='grey')
axes=ax.axis('equal')
ax.plot(x0[0],x0[1],'k.')
ax.annotate('$x_0$', xy=x0, fontsize=24, 
            xytext=(-30, 4), textcoords='offset points')
ax.axis(axes)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot_surface(xy[...,0], xy[...,1], f(xy), alpha=0.5)
ax.set_xbound([-mx, mx])
ax.set_ybound([-mx, mx])
ax.set_xlabel('$x$')
ax.set_ylabel('$y$');
```

<div class="optional">
We see the bump in the origin, which we can check has a height of $h$. For large ${\bf x}$, $f\approx x^2+\gamma^2 y^2 -y$ which has the shape of a convex quadratic function, stretched in the $x$-direction, that we have seen before. The minimum is found in the ring around the central bump where the function transitions from the bump to the quadratic. The $-y$ term causes a slight tilt in the surface, which means there is a unique minimum above the origin.

In the next plot, we show the same contours of $f$ as above but now in grey. On top we plot the contours of the quadratic model $f_{\text{quadratic}}$ around $x^{(0)}$ so that we can see the difference between the two. In addition, we plot ${\bf x}^{(1)}$ the position at the end of the first Newton step, which takes us to the minimum of the quadratic model.
</div>

```python
# initial guess x0 is chosen in cell above
fx0 = f(x0)
dfdx0 = dfdx(x0)
hessx0 = hess(x0)

# standard Newton step:
x1 = x0 - sl.solve(hessx0, dfdx0)

# quadratic model around x0
def quadratic_model(x):
    p = x-x0
    return fx0 + np.dot(p,dfdx0) + np.sum(
        np.tensordot(p, hessx0, axes=(-1,0))*p, axis=-1)/2



fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1, 1, 1)
c = ax.contour(xy[...,0], xy[...,1], f(xy), cvalues, colors='k', alpha=0.2)
ax.plot(x0[0],x0[1],'k.')
ax.annotate('$x_0$', xy=x0, fontsize=24, 
            xytext=(-30, 4), textcoords='offset points')
axes = ax.axis('equal')
minx = np.min([x0, x1, [-mx, -mx/gamma]], axis=0)
maxx = np.max([x0, x1, [mx, mx/gamma]], axis=0)
pad = (maxx-minx)*0.5
x01 = np.linspace(minx[0]-pad[0], maxx[0]+pad[0], 40)
y01 = np.linspace(minx[1]-pad[1], maxx[1]+pad[1], 40)
xy01 = np.array(np.meshgrid(x01, y01)).T
z = quadratic_model(xy01)
c = ax.contour(xy01[...,0], xy01[...,1], z, np.arange(z.min(), z.max(), 1))
ax.plot([x0[0],x1[0]],[x0[1],x1[1]],'k')
ax.plot(x1[0],x1[1],'b.')
ax.annotate('$x_1$', xy=x1, fontsize=24,
            xytext=(-30, 4), textcoords='offset points')

ax.axis(axes);

print("New position x1:", x1)
print("Achieved decrease:", f(x1)-f(x0))
print("Expected (Armijo):", np.dot(dfdx0, x1-x0))
```

<div class="optional">
If we look at the position ${\bf x}_1$ after a full Newton step, we find it to be close to the origin. This is because close to ${\bf x}_0$ the function $f$ looks a lot like a convex quadratic, but as we get further away from ${\bf x}_0$, the differences between the quadratic model $f_{\text{quadratic}}$ become larger and larger. In this particular case we have actually landed ourselves on top of the central bump, so that instead of a decrease in $f$ we have actually obtained an increase!

In a line search we would backtrack along the line segment between ${\bf x}_0$ and ${\bf x}_1$ to a point closer to ${\bf x}_0$ where we trust the quadratic model more. If we have decided we can trust the quadratic model only in the vicinity of ${\bf x}_0$ however, it seems not entirely obvious that we should still trust the search direction ${\bf x}_1-{\bf x}_0$. The idea behind <a class="definition" href="#definitions" id="trustregionmethods">trust region methods</a> is that once we have decided that ${\bf x}_1$ is too far from ${\bf x}_0$ (based on the usual sufficient decrease heuristics), we decide to only trust the quadratic model in a *trust region* closer to ${\bf x}_0$. For instance we could decide to only trust the quadratic model at a distance $\delta = \|{\bf x}_1 -{\bf x}_0\|/2$ halfway to the full Newton step $x_1$. In other words, we define a trust region

$$
  B({\bf x}_0, \delta) = \left\{{\bf x} \mid \|{\bf x} - {\bf x}_0\| \leq \delta \right\}
$$

and only use $f_{\text{quadratic}}$ within that region. The new point according to the trust region method is then given by the minimizer of $f_{\text{quadratic}}$ in $B({\bf x}_0, \delta)$.

This leads to a constrained minimisation problem (which we'll cover in lecture 10), but from the following figure, we can see:

<div class="optional">
The point ${{\bf x}_{TR}^1}\in B({\bf x}_0, \delta)$ where the minimum is achieved is where its outer circle touches a contour of $f_{\text{quadratic}}$. The normal to the circle in that point, which is in the direction ${{\bf x}_{TR}^1}-{\bf x}_0$, should be in the same direction as the negative gradient of $f_{\text{quadratic}}$ at that point:

$$
  \mu\left[{{\bf x}_{TR}^1} - {\bf x}_0\right] = -f_{\text{quadratic}}'({\bf x}_{TR}^1)
$$

where $\mu$ is a positive scalar. Using

$$
  f_{\text{quadratic}}'({\bf x}) = f'({\bf x}_0) + f''({\bf x}_0)\left[{\bf x}-{\bf x}_0\right],
$$

we can rewrite

$$
  \left[\mu\underline{\mathbf I} + f''({\bf x}_0)\right] \left[{\bf x}_{TR}^1 - {\bf x}_0\right] = - f'({\bf x}_0) \label{trust_region} \tag{trust\_region}
$$

which is a matrix equation where the matrix is formed by the sum of two parts: $\mu$ times the identity matrix, and the Hessian.

## Trust Region Methods - summary
Trust region methods limit the area in which the quadratic approximation of $f$ that underlies Newton's method is trusted 
based on a sufficient decrease criterion. It will apply Newton's method unaltered if the quadratic model is good. This will be the case if we're close enough to a local minimum, thus giving us the same local convergence as Newton's method and Newton's method with line search. If we are far away from the minimum, the quadratic model may only be accurate in the near vicinity of the current guess. In this case, rather than jumping to the minimum of the quadratic approximation, we jump to the minimum of the quadratic approximation within the trusted region. As we have shown above, this is equivalent to modifying the Hessian matrix that is used to solve for the step vector ${\bf p} = {\bf x}_{TR}^1 -{\bf x}_0$
(here ${\bf x}_0$ is the last iterate and ${\bf x}_{TR}^1$ is the next iterate computed by the trust region method)
in the following way:

$$
  f''({\bf x}_0) \to f''({\bf x}_0) + \mu\underline{\mathbf I}
$$

We consider the following two limits for $\mu$:

* for small $\mu$ (limit $\mu\to 0$) we can neglect the $\mu\underline{\mathbf I}$ part and we recover the original Newton method


* for large $\mu$ (limit $\mu\to\infty$) we can neglect the Hessian, and obtain

$$
  {\bf x}_{TR}^1 = {\bf x}_0 - \mu^{-1} f'({\bf x}_0)
$$

which is just the Steepest Descent with $\lambda=\mu^{-1}$.

In this way, trust region methods can backtrack by increasing $\mu$ (which limits the size of the trusted region) from $\mu=0$, in a similar way to line search methods that decrease $\lambda$ starting at $\lambda=1$. The line search method however will just shorten the step in the chosen Newton direction, whereas the trust region method also changes direction to slowly become more like a Steepest Descent method.

This difference will be most notable in regions where the Hessian $f''$ is not SPD. As we noted before, if $f''$ is not SPD the Newton direction is not necessarily a descent direction and it is therefore not guaranteed that a backtracking line search method will actually converge to an acceptable new point. The Steepest Descent method actually completely ignores the Hessian so is generally well-behaved. One can therefore interpret the Newton trust region method as a modification to the Hessian (by the addition of $\mu\underline{\mathbf I}$) to ensure that the modified Hessian is SPD.

To actually compute the specific $\mu$ that corresponds to a specific size of the trust region, so that $\|{\bf x}_{TR}^1-{\bf x}_0\|=\delta$, is not entirely straight-forward and can be costly. There are therefore several approaches and approximations (Dogleg, Hook, Powell's method) to find a suitable $\mu$, which we'll not go into here.

# Quasi-Newton Methods

In <a class="definition" href="#definitions" id="QuasiNewtonmethods">Quasi-Newton methods</a> rather than using the exact Hessian $f''$ (or Jacobian ${\bf F}'$), an approximation $\underline{\mathbf B}$ is used. A reason, that we have already seen, to modify the Hessian (and thus modify the exact Newton method) is to improve the *global convergence* properties. As long as we ensure that our approximation converges to the Hessian (for more precise conditions see the Dennis-Moré theorem, e.g. [Nocedal and Wright](https://link.springer.com/book/10.1007/978-0-387-40065-5)) we still have *local convergence* properties that are similar to those of pure Newton.

Another important reason to consider replacing the exact Jacobian ${\bf F}'({\bf x}^{(i)})$ (or Hessian $f''({\bf x})$ in optimisation problems) with an approximation, is that the calculation of the exact Jacobian can be very costly. After all we like to consider inversion and optimisation problems with a large number, $n$, of unknowns, so the Hessian becomes a large $n\times n$ matrix. Moreover, for a lot of applications ${\bf F}$ is not given by a simple analytical formula, and therefore no analytical derivative is available. Consider the common case where $F$ is formed by feeding ${\bf x}$ as input parameters to an expensive numerical model, and ${\bf F}({\bf x})$ are the outcomes of that model. It is not immediately obvious how one should compute the derivative of the model outcomes with respect to the input parameters. A straightforward idea is to simply perturb the input parameters one-by-one, and for each direction ${\bf p}$ approximate the derivate by

$$
  {\bf F}'({\bf x})^T{\bf p} \approx \frac{{\bf F}({\bf x}+\epsilon{\bf p})-{\bf F}({\bf x})}{\epsilon\|p\|}
  \;\;\; \text{ for small } \; \epsilon
$$

This however would require $n$ different expensive model runs, one for each of the different directions. In lecture 11 we will discuss more efficient ways of computing gradients  with respect to numerical model inputs, but it should be clear that computing the exact Hessian is not feasible in all applications.

In the rest of this lecture, and in the lecture 9 we will therefore look at methods of the following form:


* evaluate ${\bf F}({\bf x}^{(i)})$


* construct $\underline{\mathbf B}^{(i)}$, an approximation to the Jacobian $F'({\bf x}^{(i)})$


* solve $\underline{\mathbf B}^{(i)}{\bf p} = -{\bf F}({\bf x}^{(i)})$


* update ${\bf x}^{(i+1)} = {\bf x}^{(i)} + {\bf p}$



We will discuss the Gauss-Newton and Quasi-Newton secant methods (lecture 9) that fall in this category. We will also encounter Inexact Newton in lecture 9, in which we do assemble the full Hessian but only approximately solve the associated linear inversion problem.

# Newton-Gauss

A large class of inversion and optimisation problems take the form of a nonlinear least-squares problem. These are problems based on a residual function ${\bf R}:\mathbb{R}^n \to \mathbb{R}^m$ where we want to find the ${\bf x}\in\mathbb{R}^n$ that minimizes the 2-norm of the residual ${\bf R}({\bf x})$. 


For convenience we study the equivalent minimisation problem

$$
  \min_{{\bf x}} f({\bf x}), \;\;\; \text{ where } \; f({\bf x}) = \tfrac 12 {\bf R}({\bf x})^T {\bf R}({\bf x})
$$

Here the $m$ entries of ${\bf R}({\bf x})$ typically represent some equations/constraints that we may not be able to (or expect to) satisfy exactly - we just want to find the $\bf x$ for which ${\bf R}(\bf x)$ is as small as possible.

We distinguish the following cases:


* if $m\lt n$ we call the problem *underdetermined* (also underconstrained). Here we can typically find many solutions that satisfy ${\bf R}({\bf x})=0$ exactly, and these solutions may form a subset (of some dimension) of $\mathbb{R}^n$. Since they are also solutions to the minimimisation problem, one finds that applying Newton's method directly is problematic.

* if $m=n$, the *square* (*equi-determined*) case, we could also try to apply Newton's method to ${\bf R}({\bf x})=0$ directly.

* if $m>n$ the problem is overdetermined. Typically there are no exact solutions to ${\bf R}({\bf x})=0$. As we will see the magnitude of the residual in the minimum of $f$, $\|{\bf R}({\bf x}^*)\|$ plays a role in the convergence properties of the Gauss-Newton method.

In order to apply Newton's method to the minimisation of 

$$
  f = \tfrac 12 {\bf R}({\bf x})^T {\bf R}({\bf x})
$$

we first derive:

\begin{align*}
  f'({\bf x}) = {\bf R}'({\bf x})^T{\bf R}({\bf x}) \\[5pt]
  f''({\bf x}) = {\bf R}'({\bf x})^T {\bf R}'({\bf x}) + \sum_{k=1}^m R_k({\bf x}) R_k''({\bf x})
\end{align*}

<div class="optional">
Note: these expressions are more easily derived when writing out in the components $R_k$ of the residual vector $\bf R$:

$$
  f  = \sum_{k=1}^m \tfrac 12 R_k^2
$$

\begin{align*}
  \frac{\partial f}{\partial x_j} 
  &= \sum_{k=1}^m \frac{\partial \tfrac 12 R_k^2}{\partial x_j} \\
  &= \sum_{k=1}^m R_k \frac{\partial R_k}{\partial x_j} \\
  &= \sum_{k=1}^m \left[\bf R'(\bf x)\right]_{kj} R_k \\
  &= \sum_{k=1}^m \left[\bf R'(\bf x)\right]^T_{jk} R_k
  = \left[{\bf R}'(\bf x)^T {\bf R(\bf x)}\right]_j
\end{align*}

\begin{align*}
  \frac{\partial^2 f}{\partial x_i\partial x_j}
  &= \frac{\partial}{\partial x_i} \left(\sum_{k=1}^m R_k \frac{\partial R_k}{\partial x_j}\right) \\
  &= \sum_{k=1}^m \frac{\partial R_k}{\partial x_i} \frac{\partial R_k}{\partial x_j} +
     R_k \frac{\partial^2 R_k}{\partial x_i\partial x_j} \\
  &= \sum_{k=1}^m \left[{\bf R}'(\bf x)\right]_{ki} \left[\bf R'(\bf x)\right]_{kj} +
    R_k \left[R_k''(\bf x)\right]_{ij} \\
  &= \sum_{k=1}^m \left[\bf R'(\bf x)^T\right]_{ik} \left[\bf R'(\bf x)\right]_{kj} +
     R_k \left[R_k''(\bf x)\right]_{ij} \\
  &= \left[\bf R'(\bf x)^T \bf R'(\bf x)\right]_{ij} + \sum_{k=1}^m 
     R_k \left[{R_k}''(\bf x)\right]_{ij} \\
  &= \left[ {\bf R}'({\bf x})^T {\bf R}'({\bf x}) + \sum_{k=1}^m R_k({\bf x}) R_k''({\bf x}) \right]_{ij}
\end{align*}
</div>

The second term in the calculation of the Hessian $f''$ involves a sum over all $m$ components $R_k$ of the residual vector ${\bf R}$, and for each of these components we need to calculate the $n\times n$ Hessian $R_k''$. The assembly of the full Hessian needed for Newton's method therefore quickly becomes unfeasible for larger problems. The <a class="definition" href="#definitions" id="GaussNewtonmethod">Gauss-Newton method</a> therefore approximates the Hessian by leaving out the second term. The method is then given by:


* evaluate ${\bf f}'({\bf x}^{(i)}) = {\bf R}'({\bf x}^{(i)})^T {\bf R}({\bf x}^{(i)})$


* evaluate $\underline{\mathbf B}^{(i)} = {\bf R}'({\bf x}^{(i)})^T {\bf R}'({\bf x}^{(i)})$


* solve $\underline{\mathbf B}^{(i)}{\bf p} = -{\bf f}'({\bf x}^{(i)})$


* update ${\bf x}^{(i+1)} = {\bf x}^{(i)} + {\bf p}$

## Nonlinear regression

The most well-known case in this class of problems stems from nonlinear regression. Here, we want to build a model (fit a curve) $G$ that depends on $n$ model parameters (or coefficients) $x_i$ such that for a set of $m$ measurements $(X_j,  Y_j)$ we have that

$$
  Y_i \sim G(X_i; {\bf x})
$$

A least-squares fit of the model parameters ${\bf x}$ to the given data $(X_j, Y_j)$ is given by

$$
  \min_{{\bf x}} \|{\bf R}({\bf x})\|^2, \;\;\; \text{ where } \;
  R({\bf x})_j = G(X_j; {\bf x}) - Y_j, 1\leq j\leq m
$$

### The Linear Case: Linear Regression <a class="tocSkip"></a>
This section is optional, and just shows that for the linear case, Gauss-Newton comes down to the same least squares solution for linear regression that we saw in lecture 1.

<div class="optional">
Let's assume the model $G$ is simply a linear combination of some chosen interpolant functions $\phi_j$

$$
  G(X; {\bf x}) = \sum_{j=1}^n x_j {\bf\phi}_j(X)
$$

<div class="optional">
Note that $X$ plays the role of the variable in the curve $G$, whereas the $x_j$ are the coefficients that multiply the interpolant functions. As an example we could consider polynomial curve fitting where

$$
  G(X) = a_0 + a_1 X + a_2 X^2 + a_3 X^3
$$

where we have ${\bf x}=(a_0, a_1, a_2, a_3)$ and interpolant function $\phi_j(X)=X^j$ (for convenience we start the index $j$ at 0).
</div>

<div class="optional">
We want to find the coefficients $\bf x$ such that for our observations $X_i, Y_i$ we have $G(X_i, {\bf x})\approx Y_i$. Because $G(X_i, {\bf x})$ is linear in $\bf x$ we can write it using the Vandermonde matrix:
</div>

<div class="optional">
$$
  G(X_i; {\bf x}) = \underline{\mathbf G} {\bf x} \;\;\; \text{ and }\;\;\; {\bf R}({\bf x}) = \underline{\mathbf G}{\bf x} - {\bf Y}
$$

where $\underline{\mathbf G}$ is the $m\times n$ matrix:

$$
 \underline{\mathbf G} = \begin{pmatrix}
 \phi_1(X_1) & \phi_2(X_1) & \dots & \phi_n(X_1) \\
 \phi_1(X_2) & \phi_2(X_2) & \dots & \phi_n(X_2) \\
 \vdots & \vdots & \ddots & \vdots \\
 \phi_1(X_m) & \phi_2(X_m) & \dots & \phi_n(X_2)
 \end{pmatrix}
$$
</div>

<div class="optional">
It should be noted that with *linear regression*, ${\bf R}''({\bf x})={\bf 0}$, so there is no difference between Gauss-Newton and full Newton. Moreover, the function

$$
  f = \tfrac 12 {\bf R}({\bf x})^T {\bf R}({\bf x})
$$

after substitution of ${\bf R}({\bf x}) = \underline{\mathbf G}{\bf x} - {\bf Y}$

$$
  f({\bf x}) = \tfrac 12 {\bf x}^T \underline{\mathbf G}^T\underline{\mathbf G}{\bf x} - {\bf Y}^T\underline{\mathbf G}{\bf x} + \tfrac 12{\bf Y}^T{\bf Y}
$$

takes the same quadratic form as we saw in lecture 2b


$$
  f({\bf x}) = \tfrac 12 {\bf x}^T\underline{\mathbf A}{\bf x} - {\bf b}^T{\bf x} + c,
$$

with $\underline{\mathbf A}=\underline{\mathbf G}^T\underline{\mathbf G}, {\bf b}=\underline{\mathbf G}^T{\bf Y}, c=\tfrac 12{\bf Y}^T{\bf Y}$ and the unknown model parameters ${\bf x}$.

When we apply Newton to minimize any quadratic function, in the first (and only) iteration
we solve

$$
f''({\bf x^{(0)}}) {\bf p} = -f'({\bf x^{(0)}}), \text{ or }
\underline{\mathbf A}{\bf p} = {\bf b} - \underline{\mathbf A}{\bf x^{(0)}}
$$

which with ${\bf x}^{(1)}={\bf x}$ leads to

$$
  \underline{\mathbf A} {\bf x}^{(1)} = \underline{\mathbf A}{\bf x^{(0)}} +
  \underline{\mathbf A}{\bf p} = {\bf b}
$$

So that indeed in a single step we reach the solution also given by the equivalent linear system $\underline{\mathbf A}{\bf x}={\bf b}$.

The solution of the least-squares linear regression problem is therefore
exactly the normal equation we saw in lecture 1:

$$
  {\bf x} = \underline{\mathbf A}^{-1}{\bf b} = \left(\underline{\mathbf G}^T\underline{\mathbf G}\right)^{-1} \underline{\mathbf G}^T{\bf Y}
$$
</div>

A common model for population growth, is to assume that the growth is exponential:
$$
  G(X) = C_1 \exp(C_2 X)
$$
where $C_1$ and $C_2$ are two parameters to be determined, which we'll denote as a vector ${\bf x}=(C_1, C_2)$. Following the notation above we have a number of data points $X_i, Y_i$ where $X_i$ are the times (in years) and $Y_i$ the known population in those years. The residual vector then takes the form
$$
  {\bf R}({\bf x}) = \begin{pmatrix}
    G(X_1) - Y_1 \\
    G(X_2) - Y_2 \\
    \vdots \\
    G(X_m) - Y_m \\
   \end{pmatrix} = \begin{pmatrix}
    C_1 \exp(C_2 X_1) - Y_1 \\
    C_1 \exp(C_2 X_2) - Y_2 \\
    \vdots \\
    C_1 \exp(C_2 X_m) - Y_m \\
   \end{pmatrix}   
$$
    
which has a derivative

$$
  {\bf R}'({\bf x}) = \begin{pmatrix}
    \exp(C_2 X_1) & C_1 X_1 \exp(C_2 X_1) \\
    \exp(C_2 X_2) & C_1 X_2 \exp(C_2 X_2) \\
    \vdots & \vdots  \\
    \exp(C_2 X_m) & C_1 X_m \exp(C_2 X_m) \\
   \end{pmatrix}   
$$











Its second derivative is a 2x2 matrix for each of the $m$ data points:

$$
{\bf R}''({\bf x}) = 
\begin{pmatrix}
\begin{pmatrix}
  0 & X_1\exp(C_2 X_1) \\
  X_1\exp(C_2 X_1) & C_1 X_1^2 \exp(C_2 X_1)
\end{pmatrix} \\
\begin{pmatrix}
  0 & X_2\exp(C_2 X_2) \\
  X_2\exp(C_2 X_2) & C_1 X_2^2 \exp(C_2 X_2)
\end{pmatrix} \\
\vdots \\
\begin{pmatrix}
  0 & X_m\exp(C_2 X_m) \\
  X_m\exp(C_2 X_m) & C_1 X_m^2 \exp(C_2 X_m)
\end{pmatrix}
\end{pmatrix}
$$

In code:

```python
# use data defined in previous cell (rescaled)
X = (years-years[0])/(years[-1]-years[0])
Y = population/population[0]

def R(x):  # the residual function
    C1, C2 = x
    return C1*np.exp(C2*X) - Y

def dRdx(x):  # the first derivative of the residual
    C1, C2 = x
    # [:, np.newaxis] turns a flat rank-1 array of lenght m into a mx1 column array
    # with np.hstack we can stack those horizontally
    return np.hstack((np.exp(C2*X)[:,np.newaxis], (C1*X*np.exp(C2*X))[:,np.newaxis]))

def dRdx2(x):  # the second derivative of the residual
    C1, C2 = x
    m = len(X)
    # first we collect the 2x2=4 entries for each of the m data points
    # into a m x 4 array
    ans = np.hstack((
         np.zeros((m,1)),
         (X*np.exp(C2*X))[:,np.newaxis],
         (X*np.exp(C2*X))[:,np.newaxis],
         (C1*X**2*np.exp(C2*X))[:,np.newaxis]
    ))
    # then we reshape it to m x 2 x 2
    return ans.reshape((m, 2, 2))
```

Note that in this case where we only have $n=2$ model parameters, it is still feasible to compute the second derivative $R''$ of the residual, a $m\times n\times n$ array. For models that involve many more parameters (large $n$), this quickly becomes unfeasible. Let us therefore check whether we can get away with the Gauss-Newton approximation to the Hessian, which does not require a second derivative of the residual.

```python
def f(x):  # the least squares fit that we're trying to minimize
    return 0.5 * np.sum(R(x)**2)


def F(x):  # this is the 1st derivative of f: F(x) = f'(x)
    return np.dot(R(x), dRdx(x)) # = np.dot(transpose(dRdx(x)), R(x))


def full_hessian(x):  # 2nd derivative (Hessian) of f
    Rprime = dRdx(x)
    return np.tensordot(Rprime, Rprime, axes=(0,0)) + np.tensordot(
            R(x), dRdx2(x), axes=(0,0))


def gn_hessian(x):  # approximation to the Hessian according to the Gauss-Newton method
    Rprime = dRdx(x)
    return np.tensordot(Rprime, Rprime, axes=(0,0))


x0 = [0, 0]
result = sop.minimize(f, x0, jac=F, hess=full_hessian, method='trust-ncg')
print("Newton method")
print("""Succesful: {success}
Values for C1, C2: {x}
Number of iterations: {nit}
Number of function, Jacobian and Hessian evaluations: {nfev}, {njev}, {nhev}
""".format(**result))

result = sop.minimize(f, x0, jac=F, hess=gn_hessian, method='trust-ncg')
print("Gauss-Newton method")
print("""Succesful: {success}
Values for C1, C2: {x}
Number of iterations: {nit}
Number of function, Jacobian and Hessian evaluations: {nfev}, {njev}, {nhev}
""".format(**result))

plt.figure(figsize=(12,6))
plt.plot(years, Y, '.', label='Population UK (indexed at 1700)')
C1, C2 = result['x']
plt.plot(years, C1*np.exp(C2*X), label='Exponential Fit')
plt.yticks(np.arange(11))
plt.legend();
```

In the results we see that approximating the Hessian in the Gauss-Newton way still leads to convergence in a reasonable number of iterations. For both cases we have here used a trust-region method, as neither the Hessian nor the approximate Gauss-Newton Hessian are guaranteed SPD, as we can see by computing the eigenvalues:

```python
# compute eigenvalues of the full and Gauss_Newton Hessians
# if these are SPD then their eigenvalues should be strictly positive

w,v = sl.eigh(full_hessian(x0))
print("Full Hessian eigenvalue at x0:", w)

w,v = sl.eigh(gn_hessian(x0))
print("Gauss-Newton Hessian eigenvalue at x0:", w)
```

### Implementation (\*)
<div class="optional">
It is important to note that even when we approximate the Hessian in the Gauss-Newton method, we don't necessarily want to explicitly assemble that matrix to provide it to a generic implementation of the Newton iteration. We have only done so here for demonstration purposes.

The approximate Hessian matrix is used in a linear solve to compute the new search direction. If this linear solve is performed with an *iterative solver* (for instance Conjugate Gradient), all we need is to be able to perform a number of matrix-vector multiplications (typically one per iteration) with the approximate Hessian, which we can do by multiplying the vector 
with ${\bf R}$ and ${\bf R}'^T$ consecutively.
</div>

In `scipy.optimisation` the `least_squares` function provides various methods, where the `lm` method refers to Levenberg-Marquard (Gauss-Newton with trust region approach, see next section). Here we only need to provide the residual function $R({\bf x})$ and its derivative $R'({\bf x})$.

```python
x0 = [0,0]
result = sop.least_squares(R, x0, jac=dRdx, method='lm')
print("Levenberg-Marquard (Trust Region Gauss-Newton)")
print("""Succesful: {success}
Values for C1, C2: {x}
Number of function and Jacobian evaluations: {nfev}, {njev}
""".format(**result))
```

## Damped Gauss-Newton and Levenberg-Marquard

By construction, the approximate Hessian $\underline{\mathbf B}^{(i)}={\bf R}'({\bf x}^{(i)})^T {\bf R}'({\bf x}^{(i)})$ is symmetric, and semi-definite, but not necessarily positive definite (as we saw in the example of an exponential fit for population growth). For positive definiteness we need to ensure that there are *no* directions ${\bf p}$ such that

$$
  \underline{\mathbf B}^{(i)}{\bf p} = {\bf 0}
$$

which are null modes of $\underline{\mathbf B}^{(i)}$.

In lecture 1 we have learned that the rank of matrix $\underline{\mathbf B}^{(i)}$ is the same as that of ${\bf R}'({\bf x}^{(i)})$, and to prevent null modes from occurring we need that rank to be the same as $n$ (the number of columns of ${\bf R}'({\bf x}^{(i)})$. This can only be the case for the full-rank equi-determined case ($m=n$) or the full-rank over-determined case ($m>n$) as in the under-determined case $m<n$ the rank can be at most $m$.

Under this condition $\underline{\mathbf B}^{(i)}$ is SPD, and we know that this means that ${\bf p}$ is a descent direction. However as we have seen previously even if the Hessian (or its approximation) is SPD everywhere, this is not a sufficient condition for global convergence. Gauss-Newton is therefore typically combined with a line search or trust region approach. Gauss-Newton with Line Search is known as <a class="definition" href="#definitions" id="dampedGaussNewton">damped Gauss-Newton</a>, and Gauss-Newton with a trust region approach as the <a class="definition" href="#definitions" id="LevenbergMarquard">Levenberg-Marquard</a> method. With these methods global convergence can be ensured. Note however that damped Gauss-Newton still requires $\underline{\mathbf B}^{(i)}$ to be non-singular everywhere, whereas a trust region approach like Levenberg-Marquard can "repair" the approximate Hessian to remain SPD, and all we require is that ${\bf R'}({\bf x}^*)$ has full rank in the limit point ${\bf x}^*$.

As we saw in this lecture the way the (approximate) Hessian is modified in a trust region method takes the form
    
$$
  \underline{\mathbf B}^{(i)} = {\bf R}'({\bf x}^{(i)})^T {\bf R}'({\bf x}^{(i)})
  \quad\rightarrow\quad
  {\bf R}'({\bf x}^{(i)})^T {\bf R}'({\bf x}^{(i)}) + \mu\underline{\mathbf I}
$$
    
with a parameter $\mu \geq 0$. This is precisely the same modification we have seen as *damped least squares* in lecture 2a where we saw it as a type of regularisation. This means that a trust region approach applied to Gauss-Newton (Levenberg-Marquard) can be seen as applying regularisation in regions where the Hessian is not well behaved (not SPD), as we converge to the solution however we reduce $\mu$ to return to the original Gauss-Newton method.

Superlinear local convergence is only guaranteed if $\underline{\mathbf B}^{(i)}$ is a good approximation of the Hessian. If ${\bf x}^{(i)}\to {\bf x}^*$ and ${\bf R}({\bf x}^*)={\bf 0}$ (zero residual in the limit point), then local quadratic convergence will be achieved. In typical overdetermined problems however all constraints cannot be achieved simultaneously. Yet if the residual $\|{\bf R}({\bf x}^*)\|$ in the limit point is small, fairly good convergence can be achieved with damped Gauss-Newton, or Levenberg-Marquard. If the residual remains large the convergence may be slow.





## Convergence for underdetermined problems (*)

<div class="optional">
If $m\lt n$ (fewer constraints than unknowns) there are necessarily directions ${\bf p}\in\mathbb{R}^n$ for which

$$
  {\bf R}'({\bf x}^{(i)}) {\bf p} = {\bf 0},
$$

and thus $\underline{\mathbf B}^{(i)}$ becomes singular. We also know that there is no unique solution to the minimisation problem.

Instead we may demand that there are no ${\bf m}\in\mathbb{R}^m$ such that

$$
  {\bf m}^T {\bf R}'({\bf x}^{(i)}) = {\bf 0}.
$$

This is again the same condition of ${\bf R}'$ being *full rank*. The vector ${\bf m}$ represents a linear combination of the $m$ constraints, and the condition is equivalent to saying there are no linear combinations of the $m$ constraints for which the residual does not change in any direction.

Under this condition the matrix ${\bf R}'({\bf x}^{(i)}){\bf R}'({\bf x}^{(i)})^T$ is SPD (note the difference with  $\underline{\mathbf B}^{(i)} = {\bf R}'({\bf x}^{(i)})^T {\bf R}'({\bf x}^{(i)})$). Using this, a solution to the equation for ${\bf p}$

\begin{align*}
  \underline{\mathbf B}^{(i)} {\bf p} &= -f'({\bf x}^{(i)}) \\
  {\bf R}'({\bf x}^{(i)})^T{\bf R}'({\bf x}^{(i)}) {\bf p} &= -{\bf R}'({\bf x}^{(i)})^T {\bf R}({\bf x}^{(i)}) \\
\end{align*}

is given by

$$
  {\bf p} = -{\bf R}'({\bf x}^{(i)})^T \left[{\bf R}'({\bf x}^{(i)}) {\bf R}'({\bf x}^{(i)})^T\right]^{-1} {\bf R}({\bf x}^{(i)})
$$

Because $\underline{\mathbf B}^{(i)}$ is singular, this is not a unique solution but as we saw in lecture 1 and will proof in lecture 10 this is the solution that minimizes the norm of ${\bf p}$.

</div>
