## Why Computational Mathematics?

Why do we need to teach you some math, maths or mathematics?

If you look at any serious textbook or course on Machine Learning, AI, Data Science, Computational Science, Scientific Computing, Numerical Methods, Numerical Analysis, .... (basically solving problems using computers), then they will set aside significant time to cover underpinning mathematical theory.

This will include some, or often all of the following topics:
- linear algebra
- calculus
- statistics and probability
- approximation theory
- discrete algorithms (including sources of errors)
- discretisation techniques (including errors, convergence and stability)

By "serious" I mean learning material that follows the philosophy that to be competent users of computers to solve complex problems, you need an understanding of the underlying algorithms, which follow from the above mathematical topics.

By competent I mean that you know which method to use for a particular problem and why, you have a good understanding of why things may go wrong and how to improve/fix the situation, you can understand and therefore debug implementations, and if necessary you can implement your own methods if a specific method or functionality isn't immediately available to you within libraries you have access to.

## "Model"

"Model" is a term that will be very important for us. However, it is one that has an almost infinite number of meanings: conceptual model, physical model, scientific model, mathematical model, statistical model, surrogate model, numerical model, ...

Significant material on this is contained in the supplementary material notebook.

<br>

We can think of it as something that seeks to mimic a "system" or "process" that is most likely to be complex in some sense, or that we don't fully understand.

Mathematically, we can think of it as something (maybe a [function](https://en.wikipedia.org/wiki/Function_(mathematics)) )), that given some inputs provides some outputs.

<br>

>"**Philosophical considerations**\
Many types of modeling implicitly involve claims about causality. This is usually (but not always) true of models involving differential equations. As the purpose of modeling is to increase our understanding of the world, the validity of a model rests not only on its fit to empirical observations, but also on its ability to extrapolate to situations or data beyond those originally described in the model. One can think of this as the differentiation between qualitative and quantitative predictions. One can also argue that a model is worthless unless it provides some insight which goes beyond what is already known from direct investigation of the phenomenon being studied."

### All Models Are Wrong!

Before going any further with "models" it is vital to appreciate the sentiment behind the famous saying:

>"All models are wrong"

which is often expanded to:

>"All models are wrong, but some are useful"



[*Wikipedia: All models are wrong*](https://en.wikipedia.org/wiki/All_models_are_wrong)

This is an interesting topic, and it is important that you appreciate the points being made here ...

<br>

Some famous quotes from Statistician George Box (and colleagues)

- "... all models are approximations. Essentially, all models are wrong, but some are useful. However, the approximate nature of the model must always be borne in mind ..."
<br>


- "It has been said that "all models are wrong but some models are useful." In other words, any model is at best a useful fiction - there never was, or ever will be, an exactly normal distribution or an exact linear relationship. Nevertheless, enormous progress has been made by entertaining such fictions and using them as approximations."
<br>


- "All models are approximations. Assumptions, whether implied or clearly stated, are never exactly true. All models are wrong, but some models are useful. So the question you need to ask is not "Is the model true?" (it never is) but "Is the model good enough for this particular application?"
<br>

<br>

and others:

- **"Modelling in science remains, partly at least, an art. Some principles do exist, however, to guide the modeller. The first is that all models are wrong; some, though, are better than others and we can search for the better ones. At the same time we must recognize that eternal truth is not within our grasp."**
<br>


- "... it does not seem helpful just to say that all models are wrong. The very word model implies simplification and idealization. The idea that complex physical, biological or sociological systems can be exactly described by a few formulae is patently absurd. The construction of idealized representations that capture important stable aspects of such systems is, however, a vital part of general scientific analysis and statistical models, especially substantive ones, do not seem essentially different from other kinds of model."
<br>


- "A model is a simplification or approximation of reality and hence will not reflect all of reality. ... Box noted that "all models are wrong, but some are useful." **While a model can never be "truth," a model might be ranked from very useful, to useful, to somewhat useful to, finally, essentially useless.**"

## Discrete vs continuous

Some problems, systems we wish to simulate, etc, are fundamentally discrete and so well-suited to fitting onto (i.e. being solved using) computers which are also discrete. 

However, many problems we wish to solve are continuous.

For example, consider a variable that is a function of (3D) space and time - the temperature in the room you are sitting in for example.

Well whatever the answer to this is in reality, in practice if these building blocks are very small compared to the scales we are actually interested in, then we make what is called the **continuum or macroscopic assumption** (<https://en.wikipedia.org/wiki/Continuum_mechanics#Concept_of_a_continuum>) and formulate a mathematical model which does assume continuous variables!

So as an example a mathematical modeller may take a discrete system, construct a continuous scientific/mathematical model to describe it (e.g. an ordinary differential equation (ODE) or a partial differential equation (PDE)), and then a computational scientist needs to (re-)discretise it to make it finite/discrete in order to solve it on a computer!!

This may seem crazy, but it does make sense since we can't in general afford to solve a form of the underlying problem that is based on the fundamental discrete objects - we cannot afford to simulate every molecule, particle or atom in your room for example.

A key job of a numerical modeller/analyst is to decide how to represent a continuous process by something that fits on a computer, i.e. to come up with a discrete model, and to implement it in code.

Often, we represent the "continuous" world with a continuous ODE or PDE. The process of turning this into something we can consider and solve with a computer is called *discretisation*.

In doing this we need to consider many things - we will touch on some of these in this module, with the ACSE students going into more detail in the next module - Modelling & Numerical Methods.

To begin though we will keep things simple and start with some simple models that are already discrete.

## Example - the logistic map

The logistic map is a relatively simple discrete relation that can be used to model certain processes.

<br>

[The logistic map is the discrete version of the continuous logistic equation, we'll come back to this point later in this lecure.

Note that while simple, it's *nonlinear* so it's certainly not as simple as it could be - it's the nonlinearity here that leads to interesting behaviour!]

For example, consider population size and its growth.

For a start should population size be considered discrete or continuous in time?

For some biological processes this would be a continuous process (or rather assumed to be continuous even though each individual birth/death is a discrete event!), e.g. human population.

However, for others it is best ***modelled*** as being discrete - e.g. for species where generations may not overlap or where there are well-defined, discrete breeding seasons.

The logistic map is a famous discrete relation that captures this behaviour. 

It is given by

$$x_{n+1} = rx_n(1-x_n)$$

where $x_n$ is a value between zero and one

[mathematically we can write this as $x_n\in [0,1]$, with the use of square brackets indicating that we are considering the *closed* form of this [interval](https://en.wikipedia.org/wiki/Interval_(mathematics)), i.e. that the values 0 and 1 ARE included].

$x_n$ represents the ratio of the current population to the maximum possible population.  So $x$ itself can take any value (is continuous, although recall floating point representation of numbers, and see later!!), but $n$ the generation is discrete.

$r$ is a parameter, sometimes termed the *biotic potential* or the *Malthusian parameter*.

When the population size is small (and noting that $x_n$ being small implies that $x_n^2$ is very small and so can justifiably be ignored) we have

$$x_{n+1} = rx_n(1-x_n) = rx_n - rx_n^2  \approx rx_n $$

and so we can think of $r$ as being the "growth rate": 

$$r\approx x_{n+1}/x_{n}$$ 

(strictly speaking $r-1$ would generally be the quantity defined as the *growth* rate, e.g. a growth rate of 10% being equivalent to $r=1.1$. Note it doesn't really make sense with our initial assumptions on $x$ that $r$ can be negative).

If our model was simply 

$$x_{n+1} = rx_n$$ 

for all values of $x_n$ (not just small values) then we would experience unbounded [exponential growth](https://en.wikipedia.org/wiki/Exponential_growth):

$$ x_{n+1} = rx_n = r(rx_{n-1}) = \ldots = r^{n+1}x_0.$$

<br>

What the logistic map looks to do is acknowledge that with finite resources the growth rate will reduce as we approach the maximum population size.  

The simplest way to encode this limiting factor mathematically is to make the growth rate depend linearly on the population size, to be a maximum when the population size is small, and to reach zero when we are at maximum population, i.e. replace "$r$" with "$r \times (1-x_n)$", where we can interpret $x_n$ as the population size as a **proportion** or fraction of the maximum possible population - effectively we have [non-dimensionalised](https://en.wikipedia.org/wiki/Nondimensionalization) our problem. 


Simply by replacing $r$ with $r(1-x_n)$ in the discrete relation directly above is how we end up with the logistic map and non-trivial, interesting dynamics.

<br>


Note in particular that that due to 

$$\frac{x_{n+1}}{x_n} = r(1-x_n)$$

if at any point $x_n=1$, i.e. we reach that maximum population size, then all resources are extinguished and in the next generation $x_{n+1}=0$, i.e. everything dies out.

### An equivalent (but more useful for us) form of the logistic map

For reasons that will become apparent later in this lecture, instead of this classical form for the logistic map where we think of $x_n$ representing the ratio of the current population to the maximum possible population, and growth reducing (and going negative) as we approach $x_n=1$, let's instead consider the form 

\begin{align*}
x_{n+1} &= x_n + rx_n(1-x_n)\\ 
&= (r+1)x_n - r (x_n)^2
\end{align*}

which means that the "delta", i.e. the change in population satisfies

\begin{align*}
\Delta x_{n+1} := x_{n+1} - x_n &= rx_n(1-x_n)\\ 
&= r x_n - r (x_n)^2
\end{align*}

what this means is that $x_n$ is now allowed to go above one, but if it does as the RHS of this expression is now negative, the "delta" is negative, and hence the next generation will have a lower population size.  So this form effectively amounts to a rescaling of the more classical form of the logistic map seen above.

[NB. sometimes people would say "unity" rather that one - <https://math.stackexchange.com/questions/807222/is-there-a-difference-between-unity-and-1-in-applied-mathematics>]

Note also that for this form of the model, when the population size is small (and noting again that $x_n$ being small implies that $x_n^2$ is very small and so can justifiably be ignored) we have

$$x_{n+1} = (r+1)x_n - r (x_n)^2  \approx (r+1)x_n \implies \frac{x_{n+1}}{x_n} \approx r+1$$

and so we can think of $r$ as being the growth rate in the sense that growth in population size of 10% in a generation really does now equate to $r=0.1$.

Let's write a function that evaluates the right hand side (RHS) of the second form of the logistic map:

````python
def logistic_map2(r, x):
    """A function giving the RHS of the discrete logistic map (second form)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x : float
        Current population $x_n$

    Returns
    -------
    float
        The updated population $x_{n+1} = x_n + r x_n (1-x_n)$
    """
    return x + r*x*(1-x)
````

Let's generate some plots to better understand what the RHS of this relation does, i.e. what we expect the new population to be.

First, in order to figure out sensible bounds on the $x$ axis of our plotting, let's work out when the RHS of our map is zero.

The RHS can be written as 

$$-rx^2 + (1+r)x = x(1+r - rx)$$

which has two roots (values of $x$ where it is zero): $x=0$ and $1+r - rx = 0 \implies x = (1+r)/r$.

<br>

Note that if you prefer you could have used the familiar expression for the roots of a quadratic

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

from which you should also find the $x$ values of the roots as 0 and $(1+r)/r$.
    

````python
# let's plot the value of the new population for all current population sizes in between
# the bounds we just calculated, and let's consider a r value of 2
# you could check how changing this value changes the RHS

r = 2
x = np.linspace(0, (1+r)/r, 100)
y = logistic_map2(r, x)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

ax1.plot(x,y,'b',label=r'$y = x+rx(1-x)$')

ax1.set_title('The logistic map RHS version 2 ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)
````

### Expected behaviour

There are lots of ways we can start thinking conceptually, or theoretically (with a pencil and paper) about what we can do with this simple model, what behaviour we expect, how we might extend it to increase complexity (e.g. to better match a system we want to model) ...

This isn't a module on [**dynamical systems**](https://en.wikipedia.org/wiki/Dynamical_system) so we won't go into detail, but we can do some fun things to see what sort of avenues we might go down, as well as to practice our coding ...

Firstly, we can consider if our problem has any *fixed points*, i.e. in this example a population that if reached then doesn't vary in future generations. 

Mathematically, for our problem, we write this as

\begin{align*}
x_{n+1} &= x_{n} \\
\iff x_n + rx_n(1-x_n) &= x_n \\
\iff rx_n(1-x_n) &= 0 \\
\end{align*}

What two values of $x_n$ does this correspond to?

We can work this our mathematically ... [and for this form of the logistic map this is especially straightforward]

and/or we can just plot the expressions on either side of this last equation to figure this out (plotting both sides gives us two lines of course, one of which is just $y=x$, the other is our logistic map RHS we've already plotted).

````python
# let's plot the value of the new population for all current population sizes in between
# the bounds we just calculated
# Let's also plot the line y=x as the locations where these two lines intersect are out
# fixed points

r = 2
x = np.linspace(0, (1+r)/r, 100)
y = logistic_map2(r, x)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

ax1.plot(x,y,'b',label=r'$y = x+rx(1-x)$')

ax1.set_title('The logistic map RHS version 2 ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

# plot y = x line to find fixed points
# since at fixed points the next generation's population is equal to the current
ax1.plot(x,x,'k',label=r'$y = x$')

# plot the two intersection points I worked out mathematically
ax1.plot(0, logistic_map2(r,0), 'ro')
ax1.plot(1.0, logistic_map2(r,1.0), 'ro', label='intersection')
ax1.legend(loc='best', fontsize=14)
````

What does this plot tell us?

Firstly that we do indeed have two intersection points, assuming $r$ is positive. 


Ignoring the fixed point at (0,0) for now [a zero population in one generation will produce no offspring so remain at zero - this isn't a very interesting fixed point], 

if our population is below the $x$ value that corresponds to the non-zero intersection value it will clearly grow in the next generation - remember for a population given by a value on the $x$ axis, the blue line tells us the next generation's population size, with the black line the current generation size, so we can conclude this from the fact that the blue line is above the black line for $x$ values below the fixed point ($x=1$).



But if it is above that value, the next generation will be lower. This is a stable fixed point - if the population is slightly displaced from the fixed point, it will move back to the fixed point.

We can plot this evolution of population sizes in what is sometimes referred to as a [cobweb plot](https://en.wikipedia.org/wiki/Cobweb_plot).

````python
# a function to create the cobweb plot for the second form of the logistic map

def plot_logistic_map2_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Initial population, value from 0 to 1
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, (r+1)/r, 100)
    y = logistic_map2(r, x)
    ax.plot(x,y,'b',label=r'$y = x + rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    #ax.plot(1-1./r, logistic_map(r,1-1./r), 'ro', label='intersection')  
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.set_ylabel('$x + rx(1-x)$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map2(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
````

````python
# now set up a figure and run the function
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(111)

r = 0.9
x0 = 0.1
plot_logistic_map2_convergence(r, x0, 10, ax1)
````

If it's not clear what this diagram is telling us:

1. We start from an $x$ value - call it $x_0$ and plot it on the $y=x$ line of the plot - it's the green dot (this is the `x = x0` line in the cobweb plot code above).


2. Use the logistic map to find the next value which is given by the blue line - the vertical black dashed line up from the green dot indicates this operation (the `y = logistic_map2(r,x)` line in the cobweb plot code  computes the value, and `ax.plot(x, y, ... ` plots it).


3. Now we need to store this new value for use in the next iteration, i.e. we effectively set $x_1 = y$. We don't actually store all the iterations in this implementation, e.g. in a vector of length the number of iterations we are going to perform (we could easily do this), so instead we just overwrite our $x$ variable (`x=y` in the cobweb plot code) for use in the next iteration. In the diagram this operation can be visualised by the horizontal line that takes us back to the $y=x$ black line.


4. And we repeat. To help identify the iterations we change the opacity of the red dots - this will come in more useful shortly!

To change the nature of the population dynamics there are two parameters we can play with here - the initial condition $x_0$ and the parameter $r$.

Thinking about this as a generic numerical algorithm, it's vitally important that we understand/appreciate the sensitivity of our results to **both** of these parameters.

Why?

Because:

1. We might not know the starting population $x_0$ with complete accuracy - does this matter?


2. We might not know $r$ either, does this matter? Is $r$ completely knowable or will it always contain inevitable uncertainty? Here we are thinking of $r$ as some "physical" parameter that governs our system. Again note the links to COVID-19, in many ways it's $r$ that's more important to consider and get a good estimate of than the initial state $x_0$ of our system.


3. As we will see later $r$ in this simple example can also be interpreted as a numerical parameter, i.e. as a user/modeller we need to select an appropriate value - how do we do this and what are the consequences of the wrong choice?

<br>

Let's start by changing the starting value, and also compare what happens if we slightly perturb the starting value:

````python
# Let's have two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# set an r parameter
r = 0.9

# choose one starting value
x0 = 0.3
plot_logistic_map2_convergence(r, x0, 10, axs[0])

# and a second chosen by perturbing the first
pert = 0.1
plot_logistic_map2_convergence(r, x0 + pert, 10, axs[1])
````

So, for the $r$ parameter value considered here at least, we always end up at the same place - try some more starting values yourself.

For a small perturbation size the subsequent population sizes seem close - any initial difference clearly reduces to zero as we ultimately reach the same place, but do the differences decay [monotonically](https://en.wikipedia.org/wiki/Monotonic_function)?

This is essentially telling us that (for this parameter value at least), our system dynamics do not seem to be sensitive to the initial condition(s).

This could be quite important as in general we won't know the complete initial state of a system we are modelling.

### Sensitivity to model parameters

Now let's consider how much can things change, both quantitatively and qualitatively, if we change our input parameter(s) - here $r$.

````python
# let's try r = 1.9
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

r = 1.9
x0 = 0.3
plot_logistic_map2_convergence(r, x0, 20, ax1)
````

So now we are still converging towards the stable fixed point, but much much slower (i.e. taking many more populations) than we did with the parameter value above, and the convergence is no longer monotonic, we oscillate either side of the fixed point.

Now it looks like we have converged to a situation where we are "flip-flopping" between two values - our fixed point seems to have been replaced by what we might call a "mode 2" solution.

### Bifurcation diagram

So if we vary $r$ we can get quite different behaviour - both the intermediate steps as well as where we end up in the long run. 

Let's plot the ultimate behaviour as a function of the parameter $r$ in a so-called [bifurcation diagram](https://en.wikipedia.org/wiki/Bifurcation_diagram).

[NB. this implementation is [*vectorised*](https://en.wikipedia.org/wiki/Array_programming). See also <https://www.geeksforgeeks.org/vectorization-in-python/>.  If we looped over all the $r$ values this code would run much slower.]

````python
# consider N different r values - the bigger the N we select, the longer it will take!
N = 10000
r = np.linspace(1.0, 3.0, N)
# corresponding N initial x values
x = 0.1*np.ones(N)
# we're interested in long term limiting behaviour, so do quite a few iterations of the map 
iterations = 500
# but only plot the final 100 values we hit
values_to_plot = 100
# compute the logistic map dynamics for each r value, 
# and plot the final x values we hit for each r
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(111)
for it in range(iterations):
    x = logistic_map2(r,x)
    if it>(iterations-values_to_plot):
        ax1.plot(r,x,'k,',alpha=0.1)
ax1.set_title('Bifurcation diagram for the logistic map (version 2)', fontsize=16)
ax1.set_xlabel('$r$', fontsize=14)
ax1.set_ylabel('Eventual Population', fontsize=14)
````

The code starts at $x_0=0.1$ for each $r$ values, iterates the map 500 times, and plots on the $y$ axis the final 100 populations sizes (the first 400 could be called a "spin up" period). 

So on the plot, looking up from an $r$ value, distinct $y$ values being plotted tells us that the map converges to a value or values (e.g. when $r$ is less than 2 we conclude from this plot that every iteration converges to the population size of 1), whereas a "cloud" of points indicates that these 100 values are jumping around.

Note that we started every simulation using an initial condition of $x_0=0.1$. It would be wise for you to check that other starting values lead to qualitatively the same bifurcation diagram.

<br>

As $r$ increases from 1 and passes 2, the stable fixed point becomes an unstable fixed point and two "arms" appear - that is the "mode 2" solution we saw earlier, where the population jumps between these two points. As we increase $r$ further, the system eventually becomes chaotic, until it completely diverges after $r = 3$ .

# Taylor series

Hopefully most of you will have seen Taylor series before.

A Taylor series is a method by which we can approximate an arbitrary function by the sum of the function's value and the function's derivatives (appropriately weighted) at a single point, 

i.e. we achieve an approximation to the entire behaviour of the function for all input values only using knowledge about the function at a single point (althought it will only be a "good" approximation for a subset of all values).


This means we construct an approximation *around (or about or at) this point*  - we call this the "expansion point".

This concept underlies many things (especially in computational science) so it's worth us spending some time reviewing it and making sure we fully understand it.

## Motivation - a constant approximation

Consider a function of a single independent variable:  $f(x)$

Suppose this function is computationally expensive to evaluate. Suppose I have already evaluated it at a single point (our expansion point) $x_0$, i.e. I know the value of $f(x_0)$.

Given only this information, what is the best guess (estimation/approximation) I can make for the function's value at a different point $x$? I.e. how can I come up with an approximation for $f(x)$, where $x\ne x_0$

Well with no additional information I can't really do anything better than 

$$  f(x) \approx f(x_0)$$

This may seem a hopeless approximation, but it can be used in some situations. Pretend $f$ is a weather forecast and $x$ is time. This just says that for a forecast of the weather tomorrow use what we see (and therefore know) today - this is a real technique called [*persistent forecast*](https://en.wikipedia.org/wiki/Weather_forecasting#Persistence).  

"... This makes persistence a 'hard to beat' method for forecasting longer time periods": <http://ww2010.atmos.uiuc.edu/(Gh)/guides/mtr/fcst/mth/prst.rxml>

<br>

As an example, let's consider the exponential function:

````python
def f(x):
    return np.exp(x)
````

and take our expansion point $x_0$ to be $0$.

Let's plot the true function and our constant approximaton about this point:

````python
fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('A constant approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)

# define our x for plotting purposes
x = np.linspace(-3., 3., 1000)

# plot exact function
ax1.plot(x, f(x), 'k', lw=3, label='Exact function')

# define our approximation
x0 = 0.0
y = f(x0)*np.ones_like(x)
ax1.plot(x0, f(x0), 'ro', label='Expansion point')

# and plot the constant approximation
ax1.plot(x, y, 'b', lw=2, label='Constant approximation')

ax1.legend(loc='best', fontsize=14)
````

Notice that it's fairly obvious I can only have confidence in the approximation 

$$ f(x) \approx f(x_0)$$ 


if:


1. It turns out that the function $f$ is actually a constant, or


2. I can assume that $f$ doesn't vary very much in the vicinity of $x_0$, and if we are interested in $x$ values that are close to $x_0$. [How quickly $f$ varies dictating how close to the expansion point our approximation is any good].

<br>

Note that in this case the key metric for how much the function "varies" is its slope, or first derivative, at the expansion point.


<br>

In the case of the exponential function, this approximation is clearly very poor for the range of $x$ values plotted. So how can I do better?

## Motivation - a linear approximation (a "linearisation")

An obvious improvement we can make is to approximate the true function with a *linear* approximation, instead of simply a constant. 

This is a common way to write a linear function:

$$ y(x) = mx + c$$

where $m$ is the gradient (or slope) of the line and $c$ is the intercept. The value of $c$ vertically shifts the line - when $x$ is 0, $y=c$.



#### Aside - terminology

Note that "slope" and "intercept" are commonly used terms in the context of *curve-fitting*.

In the language of *(linear) regression*, especially in a *machine learning* context, we will often hear the terms "weight" and "bias".



We consider an approximation, or an expansion, (or in this case a "linearisation") "about a point".  

For us the point is $x_0$ and it makes sense for us to make the gradient of the linear approximation the same as the derivative of the function at this point.  

So we are now using the value of $f$ at $x_0$ **AND** the value of the derivative of $f$ at $x_0$:  $m = f'(x_0)$.

This is the formula for the linear line with this slope:

$$ y(x) = f'(x_0) x + c$$

We can use a further piece of information to figure out an appropriate value for the constant $c$.

We want our linearisation to pass through the function when $x = x_0$:  $y(x_0)=f(x_0)$.  

This fixes the value of $c$ for us and we have our linear approximation.

<br>

Alternatively, the simplest way to write the linearisation that satisfies this is perhaps in the following form: 


$$ y(x) = f(x_0) + (x-x_0)f'(x_0)$$


so when $x = x_0$, $(x-x_0)$ is zero and we get $f(x_0)$ on the RHS.

As we move $x$ away from $x_0$, $(x-x_0)$ grows and our approximation adds a correction based on the value of the derivative at $x_0$.


<br>


Let's plot this using the example of the exponential function. First we need to define the derivative:

````python
# of course for our example function f the derivative is trivial:
def fx(x):
    return np.exp(x)
````

````python
fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('A constant approximation', fontsize=16)
ax1.set_xlabel('$x$', fontsize=16)

# define our x for plotting purposes
x = np.linspace(-3., 3., 1000)

# plot exact function
ax1.plot(x, f(x), 'k', lw=3, label='Exact function')

# define our approximation
x0 = 0.0
y = (f(x0) + (x-x0)*fx(x0))*np.ones_like(x)
ax1.plot(x0, f(x0), 'ro', label='Expansion point')

# and plot the constant approximation
ax1.plot(x, y, 'b', lw=2, label='Linear approximation')

ax1.legend(loc='best', fontsize=14)
````

This is clearly an improved approximation compared to the constant one.

However the same accuracy points hold: if we are a long way from the expansion point, or if $f$ is very complex, this may not be a good approximation.

Now it's not the variation of the function as measured by its slope that dictates how far from the expansion point our approximation loses accuracy, now it's the second derivative or curvature.

<br>

Can we do better - yes we can consider the curvature of the black line in addition to its value and its slope, i.e. we can include a term proportional to $f''(x_0)$ to our approximation.

and so on...

This is the formula for the infinite Taylor series *about (or around) the point $x_0$* 

$$f(x) = f(x_0) + (x - x_0) f'(x_0) + \frac{(x - x_0)^2}{2!} f''(x_0) + \frac{(x - x_0)^3}{3!} f'''(x_0) +
\frac{(x - x_0)^4}{4!} f^{(iv)}(x_0) + \ldots$$

<br>

<br>

An equivalent way of writing this expansion is in  terms of the difference between $x_0$ and the perturbed point $x = x_0+h$:

$$ f(x_0+h)  = f(x_0) + hf'(x_0) + \frac{h^2}{2!}f''(x_0) + \frac{h^3}{3!}f'''(x_0) + \ldots $$

<br>

<br>

Note that it's common to see $h$ replaced by the notation $\Delta x$ or $\delta x$.

<br>

<br>


As we include more and more terms our approximation improves - see the following animated gif from Wikipedia which explains the point.

#### Taylor series example

<img src="https://upload.wikimedia.org/wikipedia/commons/6/62/Exp_series.gif" width=300x>


More terms equate with a better approximation valid a larger distance from $x_0$.

### Aside: Big O notation [\* - important concept/terminology, but in the interests of time please read through on your own]

When talking about terms in infinite series (terms we will often be forces to truncate, i.e. throw away), or talking about errors, convergence, complexity, run times etc, so-called [Big-O](https://en.wikipedia.org/wiki/Big_O_notation) is very useful.

</div>

<div class="optional">
    
As an alternative to writing "$ \ldots $" in the above infinite expansions, we can also write 

$$f(x) = f(x_0) + (x - x_0) f'(x_0) + \frac{(x - x_0)^2}{2!} f''(x_0) + \frac{(x - x_0)^3}{3!} f'''(x_0) + \mathcal{O}((x - x_0)^4)$$

or

$$ f(x_0+h) = f(x_0) + hf'(x_0) + \frac{h^2}{2!}f''(x_0) + \frac{h^3}{3!}f'''(x_0) + \mathcal{O}(h^4) $$


What does this mean?

</div>

<div class="optional">
    
#### In Mathematics

Well simply $\mathcal{O}$ here means "order".

More formally it is used to signify (place bounds on) the limiting behaviour of the magnitude of a mathematical term (or an algorithm's runtime).

So based on the the example above we have simply replaced the terms 

$$ \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots $$

with 

$$ \mathcal{O}(h^4) $$

"in the limit" (and implicitly for this use application we are interested in the limit as $h\rightarrow 0$)

i.e. we have written that

$$ \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots  =  \mathcal{O}(h^4) $$

this is stating that as $h$ tends to zero the LHS can bounded in magnitude by a term of the form

$$ C h^4 $$

where $C$ is a constant, i.e. there exists a constant $C$ such that for all sufficiently small $h$

$$\left| \frac{h^4}{4!} f^{(iv)}(x_0) +  \frac{h^5}{5!} f^{(v)}(x_0) + \ldots \right| \le C h^4 $$

In our case the point is that, assuming $f$ is a well-behaved function, i.e. it's derivatives are bounded, then there will be a small enough $h$ such that the terms dependent on the 5th and higher powers or $h$ are very small relative to the first term 

They can't be ignored, but we can select a $C$ which is a bit larger than original factor ${h^4}/{4!}$ so that the above holds.

The point here isn't to actually find $C$, this is just notation to help convey a point.

It's important when analysing errors in algorithms.

Note that for this case we were considering the limit of small $h$ and so the lowest power eventually becomes dominant, the opposite situation would occur if we were considering the limit of large $h$.

With errors we are generally thinking about the former case, with run-times (where $h$ might be replaced with some measure of the size of the problem) we are in the latter case.

</div>

<div class="optional">

#### Example

Consider

$$g(y) = 3y^2 - y^3 + 9y^4$$

in the limit as $y\rightarrow 0$.

We can show that

$$\left| 3y^2 - y^3 + 9y^4 \right| \le \left| 3y^2\right| + \left|y^3\right| + \left|9y^4 \right| \le 3y^2 + y^2 + 9y^2 \le 13 y^2$$

as we are considering the case as $y$ gets small. Hence we can write 

$$g(y) = 3y^2 - y^3 + 9y^4 = \mathcal{O}(y^2) \quad\text{as}\quad y\rightarrow 0$$


If we were considering the case of $y\rightarrow \infty$, then we would write instead

$$g(y) = \mathcal{O}(y^4) \quad\text{as}\quad y\rightarrow \infty$$

</div>


<div class="optional">
    
When dealing with numerical errors you may see things like

$$\text{error} \approx 10^{-5}\Delta t +  10^{5}\Delta t^2$$

For larger values of $\Delta t$ the second term will clearly dominate due to the relative size of the constant in front of it compared to the first term.

But when performing a convergence analysis, i.e. investigating how the error drops as $\Delta t$ is reduced, there will come a point for small enough $\Delta t$ that the first term starts to dominate, and from the point onwards the error will decay linearly rather than quadratically, i.e. halving $\Delta t$ leads to a reduction in the error by a factor of 2 rather than 4.

So we would say that 

$$\text{error} = \mathcal{O}(\Delta t)$$

even though at larger $\Delta t$ values the error would be observed to decay quadratically as we reduce $\Delta t$.

We use the expression "in the asymptotic limit" to refer to region of parameter space where the leading order behaviour dominates. For the above example if we observe something that looks like second order behaviour (when we expect first) we would explain this away by saying that we are not in the asymptotic limit. If we see very close to first order behaviour then we would say we are in the asymptotic limit.

An algorithm is said to have (*time* or *algorithmic* or *computational*) [complexity](https://en.wikipedia.org/wiki/Time_complexity) of $\mathcal{O}(n^2)$ for example,
if for large enough $n$, where $n$ is a measure of the problem size, the computational cost grows quadratically - for every doubling of the problem size the cost grows by a factor 4.

When it comes to complexity it's quite common to see things like $\mathcal{O}(n \log n)$, i.e. the algorithm scales worse than linearly, but not as bad as quadratically.

## Deriving the forward Euler scheme

An [ordinary differential equation (ODE)](https://en.wikipedia.org/wiki/Ordinary_differential_equation) is an equation containing an unknown function of a **single independent variable**, let's call it $t$, and the derivatives of that function with respect to (w.r.t.) $t$. 

It thus has the most general form:

$$ F(t, y, y',y'',...,y^{(n)}) = 0 $$

$y$ is called the *dependent variable* and $t$ the *independent variable* - the solution is written as $y\equiv y(t)$ and **we want to find or estimate the solution function $y(t)$**.

<br>

<br>


$y$ might be scalar or vector valued, we assume the former for simplicity now, for the latter we talk about a corresponding **systems of ODEs**.

If we were to have more that one (or a vector of) independent variables then we are in the world of **partial differential equations (PDEs)**.

<br>

<br>

For the purposes of deriving a simple computational (or numerical) solution method let's suppose we have an ordinary differential equation (ODE) in the simpler form

$$y' = f(t,y) \qquad\left[\text{NB. on the LHS we could also use the notation $\;\dot{y}\;$ or $\;\frac{dy}{dt}\;$}\right]
$$

This is a differential equation describing a solution function of the form $y\equiv y(t)$, i.e. the solution $y$ is a function of one variable - $t$.  


For example

$$y' = \exp(t) \qquad\text{and} \qquad y' = y$$

the first is an example where the RHS is a function of $t$ only, the second a function of $y$ only - what is the solution in both cases?

<br>

This is a more complicated example where both $t$ *and* $y$ appear on the RHS:

$$y' = y + t^3$$

<br>

The first two ($y'=y$ or $y'=\exp(t)$) are relatively easy for us to solve analytically, the final case is a bit more challenging and we may need to resort to a numerical approach (or attempt symbolic computation).

We can formulate a simple numerical solver using a truncated Taylor series.

From above we know that (assuming smoothness of the solution) the following is true:

$$ y(t+\Delta t) = y(t) + \Delta t \, y'(t) + \frac{\Delta t^2}{2!}y''(t) + \frac{\Delta t^3}{3!}y'''(t) + \mathcal{O}(\Delta t^4),$$
 
we've take an arbitrary $t$ as our expansion point.

<br>


Using the notation $y_n=y(t_n), \; y'_n=y'(t_n)$ etc, and assuming a uniform time step size $\Delta t$, this is equivalent to 

$$ y_{n+1} = y_n + \Delta t\, y'_n + \frac{\Delta t^2}{2!}y''_n + \frac{\Delta t^3}{3!}y'''_n + \mathcal{O}(\Delta t^4).$$

<br>

Dropping second-order terms (which we can justify if $y$ is "smooth", which means that $y''$ is "well-behaved", i.e. we can place a bound on its magnitude, and $\Delta t$ is chosen to be sufficiently small so that the terms containing the second AND higher powers of $\Delta t$ are relatively speaking *very small*), 

and noting that $y'_n = f(t_n,y_n)$ we are left with the discrete relation

$$y_{n+1} = y_n + \Delta t \, f(t_n,y_n).$$

<br>

This is a very famous ODE solver, or time-stepping method - termed the **forward Euler** or **explicit Euler** method.

We have turned the continuous problem into a discrete one - we have discretised. And note that this is in a form similar to our logistic map, so we know that we can write a simple piece of code that iterates over $n$ to evaluate/predict/approximate the dynamics.

Let's consider the simple model problem 

$$y' = y$$

with the initial condition

$$y(0)=1$$

and see what one step of the forward Euler method does, we will assume that we start at some $t$ from the exact solution at that time, i.e. the starting point indicated by the red dot in the following image is on the exact solution (black) line.

````python
fig = plt.figure(figsize=(8, 8))
ax1 = plt.subplot(111)
ax1.set_title('One step of Forward Euler', fontsize=16)
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$y(t)$', fontsize=16)

t = np.linspace(0, 1, 1000)

# the RHS of the ODE
def f(t,y):
    return y

# an example solution trajectory (or solution to ODE $\dot{y}=f(y), y(0)=1$)
def y(t):
    return np.exp(t)

ax1.plot(t, y(t), 'k', label='Trajectory $y=y(t)$')

# this is just some location in time
t0 = 0.35
# and an example of a time step size large enough we can see what the method is doing
dt = 0.4

ax1.plot([t0], [y(t0)], 'ro')
ax1.plot([t0, t0 + dt], [y(t0), y(t0) + dt * f(t0, y(t0))], 'b', label = r"$y'=f(t_0,y_0)$")

ax1.plot([t0 + dt], [y(t0 + dt)], 'ko', label='Exact soln at $t_0 + \Delta t$')
ax1.plot([t0 + dt], [y(t0) + dt * f(t0, y(t0))], 'go', label='FE numerical soln at $t_0 + \Delta t$')
ax1.legend(loc='best', fontsize=14);
````

The black dot in the figure above shows what the true/exact solution at the new $n$ level is, while the green dot shows what the numerical solver predicts it to be.

The difference between the two is the error that has been introduced in a single step of the method.

Now in the next step of the algorithm we of course have to start from the green dot as we assume in practice we don't know the black line and so can't start from the correct black dot.  

The next step computes

$$y_{n+1} = y_n + \Delta t \, f(t_n,y_n),$$

but note that $f$ for this particular simple test case is simply $y$. 

We assume that we don't know what the exact solution is at this time level, and hence we can't read off the correct $y$ value from the $y$ axis and use it in the RHS of our update. If we could then this would let us use the correct new slope at this time, i.e. the slope at the black dot.  

Notice that in this example this gives us a $y$ value that is too small, and hence an $f$, or slope, that is too small. Therefore our next update will give us a second point that is even more too small!

The moral of the story here is that **errors generally accumulate**, as we can see if we plot this next step of the algorithm:

````python
fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)
ax1.set_title('Two steps of Forward Euler', fontsize=16)
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$y(t)$', fontsize=16)

t = np.linspace(0, 1.5, 1000)

# the RHS of the ODE
def f(t,y):
    return y

# an example solution trajectory (or solution to ODE $\dot{y}=f(y), y(0)=1$)
def y(t):
    return np.exp(t)


ax1.plot(t, y(t), 'k', label='Trajectory $y=y(t)$')

# this is just some location in time
t0 = 0.35
# and an example of a time step size large enough we can see what the method is doing
dt = 0.4


ax1.plot([t0], [y(t0)], 'ro')
ax1.plot([t0, t0 + dt], [y(t0), y(t0) + dt * f(t0, y(t0))], 'b')
y1 = y(t0) + dt * f(t0, y(t0))
ax1.plot([t0 + dt, t0 + 2*dt], [y1, y1 + dt * f(t0+dt, y1)], 'b')

ax1.plot([t0 + dt, t0+2*dt], [y(t0 + dt), y(t0+2*dt)], 'ko')
ax1.plot([t0 + dt, t0 + 2*dt], [y(t0) + dt * f(t0, y(t0)), y1 + dt * f(t0+dt, y1)], 'go')
````

## Comparing numerical and analytical solutions qualitatively

So now we have an exact solution, and we know how to solve the ODE numerically using the forward Euler scheme.

Let's code both up and see how well they agree.

In the example below we will see what happens qualitatively if we vary the time step (i.e. now we're not checking sensitivity to any physical parameter as we did above, but here sensitivity to a numerical parameter).

````python
# don't confuse this RHS of the continuous problem
# with the RHS of the discrete map used above

def logistic_rhs(alpha, x):
    """A function giving the RHS of the continuous logistic problem

    Parameters
    ----------
    alpha : float
        A problem parameter
    x : float
        Solution value

    Returns
    -------
    float
        The RHS of the ODE
    """
    return alpha*x*(1-x)
````

````python
def x_ex(t, x0, alpha):
    """Function to evaluate the exact solution to the continuous logistic problem.

    Parameters
    ----------
    t : numpy.ndarray or float
        Times
    x0 : float
        Initial condition
    alpha : float
        A problem parameter

    Returns
    -------
    numpy.ndarray or float
        Function values for times given
    """
    return x0/(x0 + (1-x0)*np.exp(-alpha*t))
````

Note that the key lines that implement the forward Euler solver 

$$y_{n+1} = y_n + \Delta t \, f(t_n,y_n),$$

in the code below are

<br>

```Python
for n in range(len(t)-1):
    x[n+1] = x[n] + dt * logistic_rhs(alpha, x[n])
```

<br>

the rest is setting up space to store the result and plotting.

[We don't need to pass in $t$ for this problem as the independent variable does not explicitly appear in the problem RHS, but we do choose to pass in the parameter $\alpha$].

<br>

<br>

A key thing for us to appreciate is how numerical solution values can change as we alter the step size $\Delta t$. Let's evaluate the solution numerically from $t=0$ to $t=10$ using a few different $\Delta t$ values.


````python
# set some physical parameters
alpha = 1.0

# set some numerical (time-stepping) parameters
dt = 1
tend = 10

def compute_and_plot_logistic(dt, tend, alpha, x0, ax):
    """Function to plot the Forward Euler and exact solutions of the logistic
    problem on axes ax (if given)

    Parameters
    ----------
    dt : float
        Timestep size
    tend : float
        End time t_end
    alpha : float
        Parameter for exact solution
    x0 : float
        Initial population, value from 0 to 1
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # define our time levels with a fixed dt
    t = np.arange(0, tend, dt)
    # initialise an array to store all solution values (for later plotting)
    x = np.empty(len(t))
    # and set the first entry in array to the initial condition
    x[0] = x0

    # now let's do the FE time-stepping via a 'for' loop (of course other approaches possible)
    for n in range(len(t)-1):
        x[n+1] = x[n] + dt * logistic_rhs(alpha, x[n])
    ax.plot(t, x, 'b.-', label='numerical approximation')

    # let's use a finer resolution to plot the exact solution to give a smoother line.
    t_fine = np.arange(0, tend, dt/100)
    ax.plot(t_fine, x_ex(t_fine, x0, alpha), 'k', label='exact solution')
    ax.set_xlabel(r'$t$', fontsize = 16)
    ax.set_ylabel(r'$x$', fontsize = 16)
    ax.set_title('Numerical vs exact solution', fontsize = 16)
    ax.grid(True)
    ax.legend(loc='best', fontsize = 14)


# Let's have a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
# this turns the 2D array of axes labels into a 1D vector we can easily call 
axs = axs.reshape(-1)

# fix the physical parameters and the initial condition
alpha = 0.95
x0 = 0.1
tend = 100

# just vary the time step size
dt = 1    # and so alpha*dt = 0.95
compute_and_plot_logistic(dt, tend, alpha, x0, axs[0])

dt = 2    # and so alpha*dt = 1.9
compute_and_plot_logistic(dt, tend, alpha, x0, axs[1])

dt = 3    # and so alpha*dt = 2.85
compute_and_plot_logistic(dt, tend, alpha, x0, axs[2])

dt = 3.16  # and so alpha*dt = 3.002
compute_and_plot_logistic(dt, tend, alpha, x0, axs[3])
````

Based on what we've seen in this lecture, should we have expected this behaviour of our numerical ODE solver?

Yes if we do a little bit of analysis of our solver ...

Note that for this problem the solver implements:

\begin{align*}
x_{n+1} &= x_n + \Delta t \, f(t_n,x_n) \\
& = x_n + \Delta t  \alpha \, x_n(1-x_n)
\end{align*}

But notice that this takes exactly the same as the second form of our discrete logistic map:

\begin{align*}
x_{n+1} &= x_n + rx_n(1-x_n)
\end{align*}

with $r = \alpha \Delta t$ .

<br>

**We know that the behaviour of the discrete logistic map is completely dependent on the choice of parameter $r$.**

<br>

Based on this discussion and equivalence we expect our numerical solution to display identical behaviour, but now this is governed by the product $\alpha \Delta t$ - i.e. a combination of a *physical parameter*, $\alpha$, and a *numerical parameter*, $\Delta t$.

For example, if $\alpha \Delta t > 3$ we would expect our numerical simulation to blow up - and this is exactly what we see just above, and the $\Delta t$ value that leads to numerical stability/instability is completely dependent on an underlying physical parameter - **if we have a different physical parameter $\alpha$ then the $\Delta t$ choice we need to make to get a stable numerical solution changes!**

Similarly in the plots above we see that oscillatory or random (but technically stable) behaviour is seen for $\Delta t$ values that completely agree with the analysis we can carry over from our investigation of the discrete logistic map.


<br><br>

***The numerical modelling, or computational scientist, cannot select the details of the numerics independent to, or without knowledge of, the physical parameters.***

## Comparing numerical and analytical solutions quantitatively

The great thing about having an exact solution (we won't always be so lucky, and of course if we were we wouldn't have a need for numerical approximations!) is that we can compute errors quantitatively and see how they vary with our $\Delta t$ choice - we are conducting a "sensitivity study" or a "convergence analysis".

Note to avoid the fact that long term behaviour might converge to a steady state, here we only integrate up to a time we know from above is in the "unsteady" phase of the problem. When performing a convergence analysis we are doing multiple runs, some requiring lots of iterations, and hence limiting the simulation time means we conclude our analysis more rapidly.

````python
   
def time_step(alpha, x0, t0, t_max, dt):
    """Implement simple time-stepper

    Parameters
    ----------
    alpha : float
        A problem parameter
    x0 : float
        Initial condition
    t0 : float
        Initial time
    t_max : float
        The last time step
    dt : float
        Time step size

    Returns
    -------
    tuple of ndarray
        Tuple containing numpy arrays with all x and t values within
        the specified interval
    """
    x = np.array(x0)
    t = np.array(t0)
    x_all = [x0]
    t_all = [t0]  
    while t < t_max:
        # start from the most recent value
        x = x_all[-1]
        x = x + dt * logistic_rhs(alpha, x)
        x_all.append(x) 
        t = t + dt
        t_all.append(t)
    return np.array(x_all), np.array(t_all)

alpha = 0.95
t0 = 0.0
x0 = 0.1

# as this problem approaches an attractor at large times - this obfuscates or masks the early time errors
# so only run for a relatively short time
t_max = 5.

# places to store the errors we compute and the time step size values
errors = []
dt_array = []

# start from a large dt - this will only work well if the starting dt, and all
# subsequent dt values are such that we always end exactly at t_max
dt = 5.
while dt > 1.e-4:
    dt_array.append(dt)
    x_all, t_all = time_step(alpha, x0, t0, t_max, dt)
    errors.append(abs(x_all[-1] - x_ex(t_max,x0,alpha)))
    dt *= 0.5

fig = plt.figure(figsize=(16, 5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

# plot error vs dt normally
ax1.plot(dt_array, errors, 'k.', label='Error with forward Euler')
ax1.set_xlabel('$\Delta t$', fontsize=14)
ax1.set_ylabel('Error', fontsize=14)
ax1.set_title('Forward Euler convergence plot (end val)', fontsize=16)
ax1.grid(True)


# plot error vs dt with a log-log plot
ax2.loglog(dt_array, errors, 'k.', label='Error with forward Euler')
ax2.set_xlabel('$\Delta t$', fontsize=14)
ax2.set_ylabel('Error', fontsize=14)
ax2.set_title('Forward Euler convergence plot (RMS)', fontsize=16)
ax2.grid(True)

# use numpy.polyfit to find best fit linear line to data
# we expect linear asymptotic behaviour - use this parameter to discount some of  the larger dt values
start_fit = 3

line_fit1 = np.polyfit(np.log(dt_array[start_fit:]), np.log(errors[start_fit:]), 1)
ax2.loglog(dt_array[start_fit:], np.exp(line_fit1[1]) * dt_array[start_fit:]**(line_fit1[0]), 'k-', label = 'slope: {:.2f}'.format(line_fit1[0]))

# add a nice way to indicate slope
annotation.slope_marker((5.e-3, 2e-5), (1, 1), ax=ax2,
                        size_frac=0.25, pad_frac=0.05, text_kwargs = dict(fontsize = 14))

ax1.legend(loc='best', fontsize=14)
ax2.legend(loc='best', fontsize=14)
````

Some crazy things go on for too large a time step size - we see this qualitatively in the figures further above, as well as in these error plots -  for the `dt` values to the right in the plot above. 

Due to this we discard the first few values when computing the slope of the line fit (`start_fit = 3`).

Notice that the slope of the line in the log-log plot if we discard these values is very close to 1. I've added the grey triangle as a way to indicate this, and also printed the computed slope in the legend.

This is expected behaviour - the error convergence rate we expect from theory (here 1 for the forward Euler ODE solver which is expected as we threw away the second order terms and higher in our Taylor series from which we derived the solver) is only expected to hold in the "asymptotic limit".

So for smaller and smaller $\Delta t$ values we expect to get closer to a slope of 1.

A slope of 1 meaning that if we halve $\Delta t$ then this leads to exactly half the error.
