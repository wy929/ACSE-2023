## Homework - Analysing the classical form of the logistic map

Recreate the analysis of the logistic map conducted in class for the initial form of the map, i.e. with the following code for RHS function.

Specifically plot the RHS; compute analytically and plot the fixed points; investigate the sensitivity of the results to different $x_0$ and $r$ values; and plot the bifurcation diagam.

```python
def logistic_map(r, x):
    """A function giving the RHS of the discrete logistic map

    Parameters
    ----------
    r : float
        Growth rate parameter
    x : float
        Current population $x_n$, value from 0 to 1

    Returns
    -------
    float
        The updated population $x_{n+1} = r x_n (1-x_n)$
    """
    return r*x*(1-x)
```

### Solution - Classical form of the logistic map

```python
# Define a parameter value r
r = 2

# Possible population values (x) are between zero and one,
# create a vector of possible values for plotting purposes
x = np.linspace(0, 1, 100)
# 100 points should be enough to get a nice smooth line, but you 
# could increase this.
# You could also have used np.arange

# Compute the corresponding logistic_map values, as we'll be
# plotting these, let's just use variable y to store these
y = logistic_map(r, x)

# Set up a figure for plotting
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

# plot the data
ax1.plot(x,y,'b')

# add axes labels and a title, play with the fontsizes to make it look nice
ax1.set_title('The logistic map ($r={0:.3f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
ax1.set_ylabel('$rx(1-x)$', fontsize=14)
# add a grid
ax1.grid(True)
```

#### Fixed points

As we did in class for the other form of the logistic map, et's see if our problem has any *fixed points*, i.e. a population that if reached then doesn't vary in future generations. 

(Note that as discussed in the lecuture for this model $x=1$ is not a fixed point, this state implies that we will exhaust all our resources in that generation and so in the next discrete step the population goes to zero as everyone dies and no-one is born!)

Mathematically, for our problem, we write this as

\begin{align*}
x_{n+1} &= x_{n} \\
\iff rx_n(1-x_n) &= x_n
\end{align*}

What two values of $x_n$ does this correspond to?

We can work this our mathematically ... 

[$x_n=0$ and $ r(1-x_n)=1 \implies x_n = 1-1/r$]

and/or we can just plot the expressions on either side of this last equation to figure this out (plotting both sides gives us two lines of course, one of which is just $y=x$, the other is our logistic map RHS we've already plotted).

```python
# choose r = 3 this time
r = 3
x = np.linspace(0, 1, 100)
y = logistic_map(r, x)

# set up a figure and axis
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

# plot logistic map
ax1.plot(x,y,'b',label=r'$y = rx(1-x)$')
ax1.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

# plot y = x line to find fixed points
# since at fixed points the next generation's population is equal to the current
ax1.plot(x,x,'k',label=r'$y = x$')

# plot the two intersection points I worked out mathematically
ax1.plot(0, logistic_map(r,0), 'ro')
ax1.plot(1-1./r, logistic_map(r,1-1./r), 'ro', label='intersection points')
ax1.legend(loc='best', fontsize=14)
```

What does this plot tell us?

Firstly that we do indeed have two intersection points, assuming $r$ is greater than one. 

Ignoring the fixed point at (0,0) for now [a zero population in one generation will produce no offspring so remain at zero - this isn't a very interesting fixed point], 

if our population is below the $x$ value that corresponds to the non-zero intersection value it will clearly grow in the next generation - remember for a population given by a value on the $x$ axis, the blue line tells us the next generation's population size, with the black line the current generation size.



But if it is above that value, the next generation will be lower. This is a stable fixed point - if the population is slightly displaced from the fixed point, it will move back to the fixed point.

We can plot this evolution of population sizes in what is sometimes referred to as a [cobweb plot](https://en.wikipedia.org/wiki/Cobweb_plot).

```python
# first define a function to do this
def plot_logistic_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Initial population, value from 0 to 1
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 1, 100)
    y = logistic_map(r, x)
    ax.plot(x,y,'b',label=r'$y = rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
```

```python
# now set up a figure and run the function
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

r = 1.9
x0 = 0.2
plot_logistic_map_convergence(r, x0, 10, ax1)
```

If it's not clear what this diagram is telling us:

1. We start from an $x$ value - call it $x_0$ and plot it on the $y=x$ line of the plot - it's the green dot (this is the `x = x0` line in the code above).


2. Use the logistic map to find the next value which is given by the blue line - the vertical black dashed line up from the green dot indicates this operation (the `y = logistic_map(r,x)` line in the code above computes the value, and `ax.plot(x, y, ... ` plots it).


3. Now we need to store this new value for use in the next iteration, i.e. we effectively set $x_1 = y$. We don't actually store all the iterations in this implementation, e.g. in a vector of length the number of iterations we are going to perform (we could easily do this), so instead we just overwrite our $x$ variable (`x=y` in the code above) for use in the next iteration. In the diagram this operation is indicated by the horizontal line that takes us back to the $y=x$ curve.


4. And we repeat. To help identify the iterations we change the opacity of the red dots - this will come in more useful shortly!

For this example, how would we get the system to converge to the fixed point at zero? I.e, under what conditions would the unstable fixed point at zero turn into a stable fixed point?

#### Bifurcation diagram

So if we vary $r$ we can get quite different behaviour - both the intermediate steps as well as where we end up in the long run. 

Let's plot the ultimate behaviour as a function of the parameter $r$ in a so-called [bifurcation diagram](https://en.wikipedia.org/wiki/Bifurcation_diagram).

NB. this implementation is [*vectorised*](https://en.wikipedia.org/wiki/Array_programming). See also <https://www.geeksforgeeks.org/vectorization-in-python/>.  If we looped over all the $r$ values this code would run much slower.

```python
# consider N different r values - the bigger the N we select, the longer it will take!
N = 10000
r = np.linspace(2.0, 4.0, N)
# corresponding N initial x values
x = 0.1*np.ones(N)
# we're interested in long term limiting behaviour, so do quite a few iterations of the map 
iterations = 500
# but only plot the final 100 values we hit
values_to_plot = 100
# compute the logistic map dynamics for each r value, 
# and plot the final x values we hit for each r
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(111)
for it in range(iterations):
    x = logistic_map(r,x)
    if it>(iterations-values_to_plot):
        ax1.plot(r,x,'k,',alpha=0.1)
ax1.set_title('Bifurcation diagram for the logistic map', fontsize=16)
ax1.set_xlabel('$r$', fontsize=14)
ax1.set_ylabel('Eventual Population', fontsize=14)
```

The code starts at $x_0=0.1$ for each $r$ values, iterates the map 500 times, and plots on the $y$ axis the final 100 populations sizes (the first 400 could be called a "spin up" period). 

So on the plot, looking up from an $r$, distinct values tells us the map converges to a value or values, whereas a "cloud" of points indicates that these 100 values are jumping around.

<br>

As $r$ increases from 2 and passes 3, the stable fixed point becomes an unstable fixed point and two "arms" appear - that is the "mode 2" solution we saw earlier, where the population jumps between these two points. As we increase $r$ further, the system eventually becomes chaotic, until it completely diverges after $r = 4$ .

### Solution - Perturbations in the logistic map

Start with the code and example from the lecture.

```python
def logistic_map(r, x):
    return r*x*(1-x)

def plot_logistic_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the logistic map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Initial population, value from 0 to 1
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 1, 100)
    y = logistic_map(r, x)
    ax.plot(x,y,'b',label=r'$y = rx(1-x)$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The logistic map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate up to "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = logistic_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
        
# Let's have two plots side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

r=1.9
x0 = 0.7
plot_logistic_map_convergence(r, x0, 10, axs[0])
pert = 0.05
plot_logistic_map_convergence(r, x0 + pert, 10, axs[1])
```

Now write a new function which takes two starting values and outputs the difference.

```python
def compute_logistic_map_perturbations(r, x1_0, x2_0, iterations=10):
    """Function to compute size of perturbation for every iteration.

    Parameters
    ----------
    r : float
        Growth rate parameter
    x1_0 : float
        Initial population, value from 0 to 1
    x2_0 : float
        Perturbed initial population
    iterations : float, optional
        The number of iterations to compute

    Returns
    -------
    perts : numpy.ndarray
        Array of perturbation value at each iteration
    """
    # we want both our initial conditions to be between zero and one so let's check they both are
    assert (0<=x1_0<=1)
    assert (0<=x2_0<=1)
        
    # set up array and initial values for unperturbed and perturbed case
    perts = np.zeros(iterations+1)
    perts[0] = x2_0 - x1_0  
    x1 = x1_0
    x2 = x2_0
    # loop through iterations and compute difference between unperturbed and perturbed
    for it in range(iterations):
        y1 = logistic_map(r,x1)
        y2 = logistic_map(r,x2)
        perts[it+1] = y2 - y1
        # update population for next iteration
        x1 = y1
        x2 = y2
    return perts
```

Run it and plot the output for some difference starting values, perturbation sizes, and $r$ values.

```python
# plot perturbation values by iteration for r = 1.9

r = 1.9
x1_0 = 0.7
pert = 0.01
x2_0 = x1_0 + pert
perts = compute_logistic_map_perturbations(r, x1_0, x2_0, 100)

fig = plt.figure(figsize=(6, 6))
ax1 = plt.subplot(111)

ax1.set_title('Variation in perturbation with iteration', fontsize=16)
ax1.plot(range(len(perts)), perts, 'b.-')
ax1.set_xlabel('$n$', fontsize=16)
ax1.set_ylabel('$x_n$', fontsize=16)
```

```python
from math import factorial

fig = plt.figure(figsize=(7, 7))
ax1 = plt.subplot(111)

ax1.set_title('Taylor series approximation', fontsize=16)
ax1.set_xlabel('$h$', fontsize=16)
ax1.set_ylabel('Error', fontsize=16)

# exp is a convenient function to use as an example since all the derivatives are the same: df/dx = f, etc

def f(x):
    return np.exp(x)


def taylor(f, order, x):
    """Function to compute the taylor series approximation.

    Parameters
    ----------
    f : function
        Input function to be approximated by the taylor series
    order : integer
        Number of terms
    x : float
        The point the taylor series is about

    Returns
    -------
    taylor_series_expansion : numpy.ndarray
        Array of taylor series approximation at each point
    """
    taylor_series_expansion = np.zeros_like(x)
    for i in range(order):
        taylor_series_expansion += f(x0) * (x - x0)**i / factorial(i)    
    return taylor_series_expansion


# expansion point
x0 = 0.

# define some increments from x0 to test
N = 6
h = np.pi*(2.**(-1*np.arange(N)))
# above creates a numpy array with first value pi, then pi/2, then pi/4, ...

# let's plot the errors with a choice for "order" using a log-log plot
order = 4
errors = np.abs( f(x0+h) - taylor(f,order,x0+h) )

ax1.loglog(h, errors, 'b.-', label = 'error')


# Compute the slope of the line and add the corresponding straight line to show how
# well our error plot agrees
start_fit = 0

line_fit1 = np.polyfit(np.log(h[start_fit:]), np.log(errors[start_fit:]), 1)
ax1.loglog(h[start_fit:], np.exp(line_fit1[1]) * h[start_fit:]**(line_fit1[0]), 'k-', label = 'slope: {:.2f}'.format(line_fit1[0]))

ax1.legend(loc='best', fontsize=14)
```

i.e. implement and plot its behaviour for different initial conditions and different $r$ parameters, and the bifurcation diagram.

### Solution - Ricker model

```python
def ricker_map(r, x):
    return np.exp(r*(1-x))*x

# plot Ricker map for parameter r = 1.5
r = 1.5
x = np.linspace(0, 3, 100)
y = ricker_map(r, x)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(111)

ax1.plot(x,y,'b',label=r'$y = rx(1-x)$')
ax1.set_title('The ricker map ($r={0:.2f}$)'.format(r), fontsize=16)
ax1.set_xlabel('$x$', fontsize=14)
# add a grid
ax1.grid(True)

# plot fixed points
ax1.plot(x,x,'k',label=r'$y = x$')
ax1.plot(0, ricker_map(r,0), 'ro')
ax1.plot(1., ricker_map(r,1.), 'ro', label='intersection points')
ax1.legend(loc='best', fontsize=14)
```

```python
def plot_ricker_map_convergence(r, x0, iterations=10, ax=None):
    """Function to plot the iterations that the ricker map undergoes
    on axes ax (if given)

    Parameters
    ----------
    r : float
        Growth rate parameter
    x0 : float
        Starting population
    iterations : float, optional
        The number of iterations to compute
    ax : matplotlib axes, optional
        The axes to plot on
    """
    # start by plotting the underlying curves from above
    x = np.linspace(0, 3, 100)
    y = ricker_map(r, x)
    ax.plot(x,y,'b',label=r'$y = exp(r(1-x))x$')
    ax.plot(x,x,'k',label=r'$y = x$')
    ax.set_title('The ricker map ($r={0:.2f}$)'.format(r), fontsize=16)
    ax.set_xlabel('$x$', fontsize=14)
    ax.grid(True)
    ax.legend(loc='best', fontsize=14)
    # plot our starting point
    ax.plot(x0, x0, 'go') 
    # now iterate upto "iterations" and plot the path taken in (x,y) space
    x = x0
    for it in range(iterations):
        y = ricker_map(r,x)
        # plot some lines to indicate map progression
        ax.plot([x,x],[x,y],'k--',linewidth=1)
        ax.plot([x,y],[y,y],'k--',linewidth=1)
        # add a dot of increasing boldness
        ax.plot(x, y, 'ro', alpha=(it+1)/iterations)
        x=y
```

## Finding an analytical solution to the continuous logistic ODE [*]
    
Derive or verify that a solution to the logistic ODE 
    
\begin{align*}
\frac{dx}{dt} = \alpha x(1-x)
\end{align*}
    
is given by
    
$$x(t) = \frac{x_0}{x_0 + (1-x_0)e^{-\alpha t}}$$
    
where $x_0 = x(t=0)$.
   
[Hint: to derive a solution you can try to use the *separation of variables* method.  
    Verify just means take the solution I give you and through substitution check that it does indeed satisfy the ODE.]
    
</div>



### Solution - Finding an analytical solution to the continuous logistic ODE

We can use a technique called *separation of variables* to find an analytical solution

\begin{align*}
\frac{dx}{dt} &= \alpha x(1-x) \\[10pt]
\implies \int \frac{dx}{x(1-x)} &= \int \alpha\, dt\\[10pt]
&\left[ \text{We can use the fact that   } \frac{1}{x(1-x)} = \frac{1}{x}+\frac{1}{1-x} \right] \\[10pt]
\implies \int \frac{dx}{x} +  \int \frac{dx}{1-x} &= \int \alpha\, dt\\[10pt]
\implies \log(|x|) -\log(|1-x|) &= \alpha t + C\\[10pt]
\implies \log(|1-x|) - \log(|x|) &= -\alpha t - C\\[10pt]
\implies \log\left(\left| \frac{1-x}{x}\right|\right) &= -\alpha t - C\\[10pt]
\implies \left| \frac{1-x}{x}\right| &= \exp(-\alpha t - C) = \exp(-C) \exp(-\alpha t)\\[10pt]
\implies \frac{1-x}{x} &= A\exp(-\alpha t)\\[10pt]
\implies x(t) &= \frac{1}{1+A\exp(-\alpha t)}
\end{align*}

where $C$ is an arbitrary constant of integration, and through dropping the absolute value on the LHS the constant $A$ is given by $A=\pm\exp(-C)$, with a sign that we can select appropriately.

If we plug in the initial condition this allows us to fix the constant:
\begin{align*}
x(0)=\frac{1}{1+A} \implies A = \frac{1-x(0)}{x(0)}
\end{align*}

and so we have the solution

$$x(t) = \frac{x(0)}{x(0) + (1-x(0))e^{-\alpha t}}$$

We will often use the notation  $x_0 = x(t=0)$, and so write this as

$$x(t) = \frac{x_0}{x_0 + (1-x_0)e^{-\alpha t}}$$

<br>

Note that we will solve this problem (and hopefully find the same solution) using symbolic computation in a homework exercise in lecture 4.

Can you sketch what this solution looks like by hand?

Think above what happens to the $e^{-\alpha t}$ term when $t$ is zero and when it is large.
