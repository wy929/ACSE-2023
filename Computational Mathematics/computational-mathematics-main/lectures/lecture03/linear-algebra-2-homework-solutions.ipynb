## Homework - Fibonacci matrix

The Fibonacci matrix is defined as

$$Q = \begin{pmatrix}
1 & 1\\
1 & 0
\end{pmatrix}$$

Compute by hand the eigenvalues and eigenvectors of the $Q$ matrix.

Check numerically that you can use these to form a diagonalization of $Q$.

We will make use of this matrix and its diagonalisation in lecture 4.

### Solution - Fibonacci matrix

The eigenvalues of the matrix $Q$ are given by 

$$0 = |A-\lambda I| = 
\begin{vmatrix}
1-\lambda & 1\\
1 & -\lambda
\end{vmatrix} = 
\lambda^2 - \lambda -1$$

which has the roots

$$\lambda_1 = \frac{1+\sqrt{5}}{2},\qquad \lambda_2 =  \frac{1-\sqrt{5}}{2}$$

to find the eigenvectors let's solve

$$
\begin{align*}
\boldsymbol{0} & = (A-\lambda I)\boldsymbol{v}\\[10pt]
 & =
\begin{pmatrix}
1-\lambda & 1\\
1 & -\lambda
\end{pmatrix}\boldsymbol{v}
\end{align*}
$$

We could perform row operations to solve this system, but also we can notice that the solution is

$$\boldsymbol{v} = \begin{pmatrix}
\lambda\\
1
\end{pmatrix}$$

Now since our eigenvalues are distinct,

$$\boldsymbol{v}_1 = \begin{pmatrix}
\lambda_1\\
1
\end{pmatrix},\qquad
\boldsymbol{v}_2 = \begin{pmatrix}
\lambda_2\\
1
\end{pmatrix}
$$

are two linearly independent eigenvectors. We won't bother normalising them as it doesn't matter for our application.

Hence,

$$P = \begin{pmatrix}
\lambda_1 & \lambda_2\\
1 & 1
\end{pmatrix},\qquad
\Lambda =
\begin{pmatrix}
\lambda_1 & 0\\
0 & \lambda_2
\end{pmatrix}$$

Let's check this

````python
Q = np.array([[1., 1.], [1.,0.]])
lam1 = (1+np.sqrt(5))/2
lam2 = (1-np.sqrt(5))/2
v1 = np.array([lam1, 1.])
v2 = np.array([lam2, 1.])

# force v1 and v2 to be column vecs
v1.shape = (2,1)
v2.shape = (2,1)

# and stack them into the matrix P
P = np.hstack((v1,v2))
# and form the diagonal matrix with e'vals on the main diagonal
Lambda = np.diag(np.array([lam1,lam2]))

# we can check if our result is (approximately - read the docs) correct with
print('Q = P \Lambda P^{-1}:  ', np.allclose(Q , P@Lambda@(sl.inv(P))))
````

Confirm what was said in the lecture that the $P$ and thus $P^{-1}$ being orthogonal matrices means that they just correspond to simple rotations.

### Solution - transformation using the diagonalisation matrices in turn - a symmetric example

````python

A = np.array([[3.6,1.6],[1.6,3.2]])

# construct P and Lambda

# start with SciPy
lam, vecs = sl.eig(A)
# take the real components and turn the eigenvalue vector into a diag matrix
Lambda = np.diag(np.real(lam))
P = np.real(vecs)

print(Lambda)
print(P)
print('A =  P@Lambda@sl.inv(P) :', np.allclose(A, P@Lambda@sl.inv(P)))
````

First let's plot what the first multiplication

$$P^{-1}\boldsymbol{x}$$

does to the unit circle

````python
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

# construct points on a circle parametrically
theta = np.linspace(0,2*np.pi,100)
np.append(theta, 0) # add a zero at the end so we go back to the start in the plot

# A 2xn array of points
points = np.vstack( (np.cos(theta), np.sin(theta)) )

ax1.set_xlabel("$x, x'$", fontsize=14)
ax1.set_ylabel("$y, y'$", fontsize=14)
ax1.set_title('Linear transformation', fontsize=14)
ax1.grid(True)

x = np.array([1,0])
y = np.array([0,1])

# plot the circle
ax1.plot(points[0,:], points[1,:], color='blue')

# plot some vectors
ax1.quiver(x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)
ax1.quiver(y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)


# Use our P^{-1} matrix

xp = sl.inv(P)@x
yp = sl.inv(P)@y
pointsp = sl.inv(P)@points

# plot the transformed circle
ax1.plot(pointsp[0,:], pointsp[1,:], color='red')

# plot the transformed vectors
ax1.quiver(xp[0], xp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)
ax1.quiver(yp[0], yp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)

ax1.axis('equal')
````

So this rotates our blue circle and stretches it to yield the ellipse with semi-major and minor axis made up of the columns of $A$ - the blue unit vectors transform to the red vectors.

Now let's pre-multiply by $\Lambda$

````python
# before running this cell please make sure you run the previous

fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

# Use the points and the x,y from the previous cell:
x = xp
y = yp
points = pointsp

ax1.set_xlabel("$x, x'$", fontsize=14)
ax1.set_ylabel("$y, y'$", fontsize=14)
ax1.set_title('Linear transformation', fontsize=14)
ax1.grid(True)

# plot the circle
ax1.plot(points[0,:], points[1,:], color='blue')

# plot some vectors
ax1.quiver(x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)
ax1.quiver(y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)

# Now use our Lambda matrix

xp = Lambda@x
yp = Lambda@y
pointsp = Lambda@points

# plot the transformed circle
ax1.plot(pointsp[0,:], pointsp[1,:], color='red')

# plot the transformed vectors
ax1.quiver(xp[0], xp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)
ax1.quiver(yp[0], yp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)

ax1.axis('equal')
````

The blue here was the outcome of the first operation, the red the result.

So pre-multiplication by the diagonal $\Lambda$ just stretches in the $x$ and $y$ directions.

Finally we need to pre-multiply by $P$

````python
# before running this cell please make sure you run the previous

fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

# Use the points and the x,y from the previous cell:
x = xp
y = yp
points = pointsp

ax1.set_xlabel("$x, x'$", fontsize=14)
ax1.set_ylabel("$y, y'$", fontsize=14)
ax1.set_title('Linear transformation', fontsize=14)
ax1.grid(True)

# plot the circle
ax1.plot(points[0,:], points[1,:], color='blue')

# plot some vectors
ax1.quiver(x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)
ax1.quiver(y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', zorder=10)

# Now use our P matrix

xp = P@x
yp = P@y
pointsp = P@points

# plot the transformed circle
ax1.plot(pointsp[0,:], pointsp[1,:], color='red')

# plot the transformed vectors
ax1.quiver(xp[0], xp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)
ax1.quiver(yp[0], yp[1], angles='xy', scale_units='xy', scale=1, color='r', zorder=10)

ax1.axis('equal')
````

Again the blue was the output from the previous operations, and red the output from the final pre-multiplication.

So again as the matrix is orthogonal the final multiplication simply rotates our previous result.

Note that (as long as you run the three cells in order!) the red result here agrees with the transformation under the matrix $A$.

## Homework $2\times 2$ ODE example from class

From the lecture consider the $2\times 2$ problem


$$\dot{\boldsymbol{x}}(t) = A\boldsymbol{x}$$

where

$$\boldsymbol{x}(t) = 
\begin{pmatrix}
x_1(t)\\
x_2(t)
\end{pmatrix}$$

and 

$$A = 
\begin{pmatrix}
-1 & 3 \\
3 & -1
\end{pmatrix}
$$

Show that the solution is given by

\begin{align*}
\boldsymbol{x}(t) &= \exp(tA) \boldsymbol{C}
 = \frac{1}{2}(C_1+C_2)
\begin{pmatrix}
1 \\
1
\end{pmatrix}\exp(2t)+
\frac{1}{2}(C_2-C_1)
\begin{pmatrix}
-1 \\
1
\end{pmatrix}\exp(-4t)
\end{align*}

Hint: You will need to start by the eigenvalues and eigenvectors of $A$.

For the case with initial condition


$$\boldsymbol{x}(0)=
\begin{pmatrix}
0.1 \\
0
\end{pmatrix}
$$

compute and plot the exact solution as well as a numerical solution using forward Euler over the interval $t\in[0,1]$.

You should find that the solution(s) tend(s) to $\pm$ infinity for large times, depending on the signs of the initial conditions.

Looking at the form of the solution, what initial conditions lead to a solution which tends to zero as time increases? Does the numerical solution agree with this qualitative behaviour (especially in the case of selecting too large a time step size)?

### Solution $2\times 2$ ODE example from class

First we need to compute the eigenvalues:

To do this we know that we need to solve the equation

$$ \det (A - \lambda I) = 0$$

for its roots $\lambda$:

$$ 0 =  \det (A - \lambda I) = 
\left|
  \begin{array}{rr}
    -1-\lambda & 3 \\
    3 & -1-\lambda  
  \end{array}
\right|
= (-1-\lambda)^2 - 9 = \lambda^2 + 2\lambda - 8 = (\lambda-2)(\lambda+4)
$$

so the eigenvalues are

$$\lambda_1 = 2\quad\text{and}\quad \lambda_2 = -4$$

Let's check our solution:


````python
A = np.array([[-1,3],[3,-1]])
print(sl.eigvals( A ))
# and if we know that the eigenvalues are real we can extract the real part:
print(np.real(sl.eigvals( A )))
````

Now let's find the eigenvectors.

Consider the first eigenvalue $\lambda_1 = 2$.

We want to find the vector $\boldsymbol{v}$ such that

$$A \boldsymbol{v} = \lambda \boldsymbol{v} \iff (A-\lambda I) \boldsymbol{v} =\boldsymbol{0}$$

which for our current example reads

$$
\left(
  \begin{array}{rr}
    -1 & 3 \\
    3 & -1  
  \end{array}
\right)
\left(
  \begin{array}{r}
    v_1 \\
    v_2  
  \end{array}
\right)
  = 2
  \left(
  \begin{array}{r}
    v_1 \\
    v_2  
  \end{array}
\right)
$$

or 

$$
\left(
  \begin{array}{rr}
    -3 & 3 \\
    3 & -3   
  \end{array}
\right)
\left(
  \begin{array}{r}
    v_1 \\
    v_2  
  \end{array}
\right)
  = 
  \left(
  \begin{array}{r}
    0 \\
    0  
  \end{array}
\right)
$$

This clearly has the solution $v_1=v_2$ 


$$\boldsymbol{v}_1 =   \left(
  \begin{array}{r}
    {1} \\
    {1} 
  \end{array}
\right)
$$


Now previously we normalised the eigenvector, but actually it doesn't really matter in this case as we are diagonalising the matrix and so multiplying by the matrix of eigenvectors and its inverse - the constants we would introduce to normalise would just cancel, so let's not bother here.

For the second eigenvector we can repeat the above with $\lambda=-4$, alternatively we can remember that the eigenvectors of a symmetric matrix are orthogonal and immediately write down

$$\boldsymbol{v}_2 =   \left(
  \begin{array}{r}
    {-1} \\
    {1} 
  \end{array}
\right)
$$

Let's check against SciPy (these will be normalised)

````python
A = np.array([[-1,3],[3,-1]])
lam, vecs = sl.eig(A)
print(vecs[:,0])
print(vecs[:,1])
````


From the lecture the solution is given by

\begin{align*}
\boldsymbol{x}(t) &= \exp(tA) \boldsymbol{C}\\[5pt]
\end{align*}

and from lecture we also know that

\begin{align*}
\exp(tA) &= 
 P\begin{pmatrix}
e^{\lambda_1 t} & 0 &  \cdots & 0 \\
0 & e^{\lambda_2 t} &  \ddots & 0\\
\vdots & &  \ddots &  \vdots \\
0 & \cdots &  0 & e^{\lambda_n t}
\end{pmatrix}
 P^{-1}
\end{align*}

Where $P = \left[ \; \boldsymbol{v_1} \;\;  \boldsymbol{v_2}  \; \; \cdots \; \;  \boldsymbol{v_n} \; \right]$ .

In our example here, the solution is then given by

\begin{align*}
\boldsymbol{x}(t) &= 
\begin{pmatrix}
1 & -1 \\
1 & 1
\end{pmatrix}
\begin{pmatrix}
e^{2t} & 0\\
0 & e^{-4t}
\end{pmatrix}
\begin{pmatrix}
1/2 & 1/2\\
-1/2 & 1/2
\end{pmatrix}
\begin{pmatrix}
C_1\\
C_2
\end{pmatrix}\\[5pt]
& = \frac{1}{2}(C_1+C_2)
\begin{pmatrix}
1 \\
1
\end{pmatrix}\exp(2t)+
\frac{1}{2}(C_2-C_1)
\begin{pmatrix}
-1 \\
1
\end{pmatrix}\exp(-4t)
\end{align*}

Let's plot this, along with a numerical solution using the initial condition

$$\boldsymbol{x}(0)=
\begin{pmatrix}
0.1 \\
0
\end{pmatrix}
$$

````python
# in order to reuse the code from the MSD example I will
# assume that the two components of the system are labelled
# x and v

# initial data (we need two for this coupled problem)
x0 = 0.1
v0 = 0
# simulation period and time step size
tend = 1
dt = 0.01


# the RHS matrix which defines the problem
A = np.matrix([[-1, 3], [3, -1]])
# the 2x2 identity matrix which we also need in our rearranged time-stepping expression
I = np.identity(2)
# set up some vectors of times and to store solution
t = np.arange(0, tend, dt)
y = np.empty((len(t), 2))
# initialise solution
y[0, :] = np.array((x0, v0))
# time step
for n in range(len(t)-1):
    y[n + 1, :] = (I + A * dt) @ y[n, :]

# pull out the individual solution vectors (don't need to do this)
x = y[:, 0]
v = y[:, 1]

fig = plt.figure(figsize=(8, 6))
ax1 = plt.subplot(111)
ax1.plot(t, x, 'b', label='$x$')
ax1.plot(t, v, 'r', label='$v$')
ax1.set_xlabel(r'$t$', fontsize=14)
ax1.set_ylabel(r'$x,\; v$', fontsize=14)
ax1.set_title('Numerical vs exact solution', fontsize=16)
ax1.grid(True)


def exact_sol(t,x0):
    x_exact = np.zeros((len(t),2))
    v1 = np.array([1,1])
    v2 = np.array([-1,1])
    # force v1 and v2 to be column vecs
    v1.shape = (2,1)
    v2.shape = (2,1)
    x_exact = 0.5*(x0[0]+x0[1])*np.exp(2*t)*v1 + \
              0.5*(x0[1]-x0[0])*np.exp(-4*t)*v2
    return x_exact

tfine = np.arange(0, tend, dt)
x_exact = exact_sol(tfine,np.array([x0,v0]))
ax1.plot(tfine, x_exact[0,:], 'b:', label='exact $x_1$ solution')
ax1.plot(tfine, x_exact[1,:], 'r:', label='exact $x_2$ solution')

ax1.legend(loc='best', fontsize=14);
````

OK, looks OK, but note that for these initial conditions both the exact and the numerical solution go off to infinity - so both are *unstable*.  

That's fine. What should concern us is a situation where the exact solution is stable but the numerical solution is unstable! Let's try to investigate this scenario for this example.

Looking at the form of the solution, in order for the solution to tend to zero (i.e. to be stable) we need to kill off the $\exp(2t)$ term. The coefficient of this is $(C_1+C_2)/2$, so choosing the two initial conditions to be minus one anohter, e.g. $1$, $-1$, should do it.  But what happens if we run with too large a time step?

Run the code with `dt=0.1` and `dt=1` and note the difference in behaviour.

## Homework - eigenvalues of the MSD system

Find the eigenvalues of the matrix that corresponds to the MSD matrix from the lecture:

$$
\begin{pmatrix}
0 & 1\\
-\omega_0^2 & -2\zeta\omega_0
\end{pmatrix}.$$

### Solution - eigenvalues of the MSD system


We can find the eigenvalues for our MSD system by finding the roots of the [characteristic equation](https://en.wikipedia.org/wiki/Eigenvalues_and_eigenvectors#Eigenvalues_and_the_characteristic_polynomial), i.e. solving the algebraic equation

$$\det(A - \lambda I) = 0,$$

for $\lambda$.

<br>

Let's do this for the $A$ for the MSD problem:

$$\det(A - \lambda I) = 
\begin{vmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{vmatrix}
= \lambda^2 + 2\zeta\omega_0 \lambda + \omega_0^2. $$


The eigenvalues are given by the roots of this quadratic. Recall that a formula for the two roots of a quadratic in the form $ax^2+bx+c=0$ is

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

Applying this in our case yields the two eigenvalues

$$\lambda_1 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right),\;\;\;\;
\lambda_2 = \omega_0\left(-\zeta - \sqrt{\zeta^2-1}\right).$$


Note that we have three cases:


1. $\zeta=0$: purely imaginary eigenvalues


2. $0<\zeta<1$: complex eigenvalues with $Re(\lambda)<0$


3. $\zeta\ge 1$: real (negative) eigenvalues



$\zeta$ here is encoding something about the physics of our problem (refer to the linked videos above and <https://en.wikipedia.org/wiki/Harmonic_oscillator#Damped_harmonic_oscillator>).

In the three cases above the system is termed: *undamped*, *under-damped* and *over-damped* (with $\zeta=1$ *critically-damped*), respectively.

### Solution - eigenvectors of the MSD system

To find the corresponding eigenvectors we need to solve the system

$$
\boldsymbol{0} = (A - \lambda I)\boldsymbol{v} = 
\begin{pmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{pmatrix}\boldsymbol{v}
$$

where 

$$\lambda_1 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right),\;\;\;\;
\lambda_2 = \omega_0\left(-\zeta - \sqrt{\zeta^2-1}\right).$$

Consider $\lambda_1$

$$
\begin{align*}
\boldsymbol{0} & =
\begin{pmatrix}
-\lambda & 1\\
-\omega_0^2 & -2\zeta\omega_0 - \lambda
\end{pmatrix}\boldsymbol{v}\\[10pt]
& =
\begin{pmatrix}
-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1\\
-\omega_0^2 & -2\zeta\omega_0 -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}\boldsymbol{v}\\[10pt]
& =
\begin{pmatrix}
-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1\\
-\omega_0^2 & -\omega_0 \left(\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}\boldsymbol{v}
\end{align*}
$$

Let's form the augmented matrix and use row operations/Gassian elimination to solve this problem


\begin{align*}
[A \, | \, \boldsymbol{b}] = 
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0^2 & -\omega_0 \left(\zeta + \sqrt{\zeta^2-1}\right) & 0 
  \end{array}
\right]\\[10pt]
\xrightarrow{(2)\leftarrow (2)\times \left(-\zeta + \sqrt{\zeta^2-1}\right)/\omega_0}
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & -\left(\zeta + \sqrt{\zeta^2-1}\right)\left(-\zeta + \sqrt{\zeta^2-1}\right) & 0
  \end{array}
\right]\\[10pt]
= & \left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0
  \end{array}
\right]\\[10pt]
& \qquad\text{[as $-\left(\zeta + \sqrt{\zeta^2-1}\right)\left(-\zeta + \sqrt{\zeta^2-1}\right)= \zeta^2 - (\zeta^2-1) = 1$]}\\[10pt]
\xrightarrow{(2)\leftarrow (2)-1}
&
\left[
  \begin{array}{cc|c}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1 & 0  \\
   0 & 0 & 0
  \end{array}
\right]
\end{align*}

This shows that this system only has one linearly independent row, hence the original matrix ($A-\lambda I$) is singlular. Of course this is exactly as expected based on how we define our eigenvalues in the first place, and so for this $2\times 2$ problem we could have just considered one of the rows immediately without going through these row operations, for larger problems however we would need to do this in general.

What this exercise allows us to do is find $\boldsymbol{v}$ - the same $\boldsymbol{v}$ solves the original problem
$A\boldsymbol{v}=\lambda \boldsymbol{v}$ as solves the transformed problem encoded in the augmented system above:

$$
\left[
  \begin{array}{cc}
    -\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) & 1   \\
   0 & 0
  \end{array}
\right]
\boldsymbol{v} = \boldsymbol{0}$$

This tells us that 

$$-\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) v_1 + v_2 = 0
\iff v2 = \omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right) v_1 
$$

we therefore have a family of possible solutions which can be written as

$$\boldsymbol{v} = \alpha
\begin{pmatrix}
1\\
\omega_0\left(-\zeta + \sqrt{\zeta^2-1}\right)
\end{pmatrix}$$

for any scalar $\alpha$.  We can choose this to make this a unit vector:

$$\alpha = \frac{1}{\sqrt{1+\omega_0^2\left(-\zeta + \sqrt{\zeta^2-1}\right)\overline{\left[\left(-\zeta + \sqrt{\zeta^2-1}\right) \right]}}}
$$
where the overline indicates complex conjugate.

### Solution - stability of forward Euler applied to the MSD problem

Here we implement a solver and a function for MSD. First we check it gives the same plot as in the lecture (note the "eyeball norm" is a weak norm to use for code verification and you should do this more rigorously by directly comparing code output).

Then we see what happens in the cases of the physical/numerical parameters given in the lecture.

````python
def forward_euler(f, u0, t0, t_max, dt):
    """ Forward Euler time-stepper.

    Parameters
    ----------
    f : callable
        f(t,y) is the RHS function
    u0 : array_like
        The initial condition
    t0 : float
        The initial time
    t_max : float
        The end time
    dt : float
        Time step size.

    Returns
    -------
    tuple of ndarray
        Tuple containing ndarrays with all u and t values
    """
    u = np.array(u0)
    t = np.array(t0)
    u_all = [u0]
    t_all = [t0]
    while t < t_max:
        u = u + dt*f(t, u)  # euler guess
        u_all.append(u)
        t = t + dt
        t_all.append(t)
    return np.array(u_all), np.array(t_all)

def f_msd(t, u):
    """RHS function for mass-spring-damper problem.

    Parameters
    ----------
    t : float or array_like
        Time
    u : array_like
        For MSD u is the vector [x, v]

    Returns
    -------
    f : ndarray
        RHS function value
    """
    f = np.zeros_like(u)
    f[0] = u[1]
    f[1] = -2.*zeta*om0*u[1] - om0**2 * u[0]
# NB. could write this as A*u with
# A = np.matrix([[0, 1], [-om0**2, -2*zeta*om0]])
    return f

# Let's first check this against the implementation from class - eyeball norm a plot

# numerical parameters for problem
om0 = 1
zeta = 0.1
# initial data (we need two for this coupled problem)
x0 = 1
v0 = 0
# simulation period and time step size
t0 = 0.
tf = 20.
dt = 0.04

# initial condition
u0 = np.array([x0, v0])
uall, tall = forward_euler(f_msd, u0, t0, tf, dt)


# pull out the individual solution vectors (don't need to do this)
x = uall[:, 0]
v = uall[:, 1]

fig = plt.figure(figsize=(8, 6))
ax1 = plt.subplot(111)
ax1.plot(tall, x, 'b', label='x')
ax1.plot(tall, v, 'r', label='v')
ax1.set_xlabel(r'$t$', fontsize=14)
ax1.set_ylabel(r'$x,\; v$', fontsize=14)
ax1.set_title('Numerical vs exact solution', fontsize=16)
ax1.grid(True)


def msd_exact_sol(t, zeta, om0):
    """Exact solution to the general MSD problem 

    Parameters
    ----------
    t : array_like
        Times
    zeta : float
        Numerical parameter zeta
    om0 : float
        Numerical parameter omega_0

    Return
    ------
    tuple of ndarray
        x and v component of solution
    """
    om = om0 * np.sqrt(1-zeta**2)
    phi = np.arctan2( (-v0 - om0*zeta*x0) , (om*x0) )
    CC = x0/(2*np.cos(phi))
    a = -om0*zeta
    b = om0 * np.sqrt(1-zeta**2)
    x_exact = 2*CC*np.exp(a*t)*np.cos(b*t+phi)
    v_exact = 2*CC*np.exp(a*t)*(-b*np.sin(b*t+phi) + a*np.cos(b*t+phi))
    return x_exact, v_exact


tfine = np.arange(0, tf, dt/100)
x_exact, v_exact = msd_exact_sol(tfine, zeta, om0)
ax1.plot(tfine, x_exact, 'k', label='exact x solution')
ax1.plot(tfine, v_exact, 'g', label='exact v solution')
ax1.legend(loc='best', fontsize=14);
````

### Solution - the double pendulum

We will use the MSD template from the lecture for this.

````python
from scipy.integrate import solve_ivp

# we should really pass in the parameters rather than hard-coding them
def f_Double_Pendulum(t, state):
    """RHS vector for the double pendulum system

    Parameters
    ----------
    t : float or array_like
        Time
    state : array_like
        State vector [th1, th2, p1, p2]

    Returns
    -------
    list of float
        [th1_dot, th2_dot, p1_dot, p2_dot]
    """
    # define problems parameters:
    g = 9.8
    m = 1  # assume both masses the same
    L = 1  # assume both lenths the same
    # unpack state
    th1, th2, p1, p2 = state
    # derivatives
    th_diff = th1 - th2
    c_th_diff = np.cos(th_diff)
    s_th_diff = np.sin(th_diff) 
    mL2 = m*L**2
    th1_dot = 6/mL2 * ( 2*p1 - 3*c_th_diff*p2) / ( 16 - 9*c_th_diff**2 )
    th2_dot = 6/mL2 * ( 8*p2 - 3*c_th_diff*p1) / ( 16 - 9*c_th_diff**2 )
    p1_dot = -0.5*mL2 * ( th1_dot*th2_dot*s_th_diff + 3*(g/L)*np.sin(th1))
    p2_dot = -0.5*mL2 * ( -th1_dot*th2_dot*s_th_diff + (g/L)*np.sin(th2))
    return [th1_dot, th2_dot, p1_dot, p2_dot]


# Define initial conditions
th1 = 180*np.pi/180 # convert degree to radians
th2 = 170*np.pi/180 # 180 points straight up
p1 = 0.
p2 = 0.

# Integration window
t0 = 0.
tf = 10.
# output time we want from the sovler
ts = np.linspace(t0, tf, 200)


# A solver tolerance - not really needed here, use default
tol = 1.e-5
# call the solver
sol = solve_ivp(f_Double_Pendulum, [t0, tf], np.array([th1, th2, p1, p2]), method='RK45', t_eval=ts, rtol=tol, atol=tol)

fig, ax1 = plt.subplots(1, 1, figsize=(8, 5))
ax1.plot(ts, sol.y[0,:], 'b', label='x1')
ax1.plot(ts, sol.y[1,:]-sol.y[0,:], 'k', label='x2-x1')
ax1.set_xlabel('$t$', fontsize=16)
ax1.set_ylabel('$\theta(t)$', fontsize=16)
ax1.set_title('Coupled pendulum problem (Angles)', fontsize=16)
ax1.legend(loc='best', fontsize=16)
````

````python
# let's generate a movie
from matplotlib import animation
from IPython.display import HTML

def pendulum_get_xy_from_angles(th1, th2):
    """Convert the two angles describing this problem
    into the x and y locations of the two end points of
    the pendula. Remember that the angles point to the 
    midpoint of the rods making up the pendula.

    Parameters
    ----------
    th1 : float
        Angle between first rod and vertical in radians
    th2 : float
        Angle between second rod and vertical in radians

    Returns
    -------
    tuple of float
        xy coordinates of the rods
    """
    L = 1.
    x1 = (L)*np.sin(th1)
    y1 = -(L)*np.cos(th1)
    x2 =  x1 + 2*(L*(np.sin(th1) + 0.5*np.sin(th2)) - x1)
    y2 =  y1 + 2*(-L*(np.cos(th1) + 0.5*np.cos(th2)) - y1)
    return x1, y1, x2, y2


# Define initial conditions
th1 = 180*np.pi/180
th2 = 170*np.pi/180
p1 = 0.
p2 = 0.

# Integration window
t0 = 0.
tf = 20.
# output time we want from the sovler
ts = np.linspace(t0, tf, 500)

# A solver tolerance - not really needed here, use default
tol = 1.e-5
# call the solver
sol = solve_ivp(f_Double_Pendulum, [t0, tf], np.array([th1, th2, p1, p2]), 
                method='RK45', t_eval=ts, rtol=tol, atol=tol)

th1 = sol.y[0,:]
th2 = sol.y[1,:]

fig = plt.figure(figsize=(7, 7))
ax1 = fig.add_subplot(111)
ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-2.5, 2.5)

# initialise the "lines" making up the moving part of the animation
line0 = ax1.plot([], [], 'k', lw=3)
line1 = ax1.plot([], [], 'k', lw=3)
line2 = ax1.plot([], [], 'ro')
line3 = ax1.plot([], [], 'bo')  
# combine the line objects - these are the things we will update between anin frames
lines = line0 + line1 + line2 + line3
plt.close()

def init():
    for line in lines:
        line.set_data([],[])
    return lines

def data_gen(framenumber, th1, th2):
    x1, y1, x2, y2 = pendulum_get_xy_from_angles(th1[framenumber], th2[framenumber])
    lines[0].set_data([0, x1], [0, y1])
    lines[1].set_data([x1, x2], [y1, y2])
    lines[2].set_data(x1, y1)    
    lines[3].set_data(x2, y2)    
    return lines

anim = animation.FuncAnimation(fig, data_gen, frames=np.arange(0, len(th1)), fargs=(sol.y[0,:], sol.y[1,:]),
                               interval=50, blit=True)

# save the animation as gif or mp4 if needed
#anim.save('animations\double_pendulum.gif', writer='imagemagick')
#anim.save('animations\double_pendulum.mp4', extra_args=['-vcodec', 'libx264'])
# display in notebook
#HTML(anim.to_jshtml())
HTML(anim.to_html5_video())
````

## Homework - when does the SVD agree with the eigen-decomposition? [\*]

Consider a square matrix - check whether the SVD and the eigen-decompositions agree. 

Compare the cases of a symmetric matrix and a non-symmetrix matrix.
    
</div>

### Solution - when does the SVD agree with the eigen-decomposition?

````python
# first the eigendecompositon of a symmetric matrix
A = np.array([[3., 1.], [1.,3.]])
lam, vecs = sl.eig(A)
print('\nlam = ')
pprint(lam)
print('\nvecs = ')
pprint(vecs)
# construct a Lambda matrix from the eigenvalues
Lambda = np.diag(np.real(lam))
print('\nLambda = ')
pprint(Lambda)
print('\nCheck our eigendecomposition works = ')
print('A =  vecs@(Lambda@sl.inv(vecs)) :', np.allclose(A, vecs@(Lambda@sl.inv(vecs))))
# and as vecs should be orthogonal for a symmetric matrix
print('\nCheck our eigendecomposition works using a transpose rather than inverse = ')
print('A =  vecs@(Lambda@(vecs.T)) :', np.allclose(A, vecs@(Lambda@(vecs.T))))
````

````python
# now a non-symmetric matrix
A = np.array([[2.2,0.3],[1.7,2.5]])
lam, vecs = sl.eig(A)
print('\nlam = ')
pprint(lam)
print('\nvecs = ')
pprint(vecs)
# construct a Lambda matrix from the eigenvalues
Lambda = np.diag(np.real(lam))
print('\nLambda = ')
pprint(Lambda)
print('\nCheck our eigendecomposition works = ')
print('A =  vecs@(Lambda@sl.inv(vecs)) :', np.allclose(A, vecs@(Lambda@sl.inv(vecs))))
# and as vecs should be orthogonal for a symmetric matrix
print('\nCheck our eigendecomposition works using a transpose rather than inverse = ')
print('A =  vecs@(Lambda@(vecs.T)) :', np.allclose(A, vecs@(Lambda@(vecs.T))))
````

So in this case, as long as we take the inverse (rather than the transpose), the eigen-decomposition works.

Now let's see what SVD does for these two matrices

````python
# repeat the symmetric case
A = np.array([[3., 1.], [1.,3.]])

# Form the SVD
UU, SS, VVT = sl.svd(A)
SSS = np.zeros_like(A, dtype='float')
# place the entries of SS into a matrix of the correct size
SSS[:min(A.shape[0],A.shape[1]),:min(A.shape[0],A.shape[1])] = np.diag(SS)

print('\nU = ')
pprint(UU)
print('\nS = ')
pprint(SSS)
print('\nV = ')
pprint(VVT.T)
print('\nOur SVD is equal to A: ',np.allclose(A, UU @ SSS @ VVT))
````

So notice that subject to minus signs (as eigenvectors are only defined upto an arbitrary scaling, so even if normalised they're only defined up to a $\pm 1$ factor) this is equivalent to the above eigendecomposition, i.e. the SVD has recovered the eigenvalues and eigenvectors of $A$ as the singular values and both the left and the right singular vectors.

````python
# repeat the nonsymmetric case
A = np.array([[2.2,0.3],[1.7,2.5]])

# Form the SVD
UU, SS, VVT = sl.svd(A)
SSS = np.zeros_like(A, dtype='float')
# place the entries of SS into a matrix of the correct size
SSS[:min(A.shape[0],A.shape[1]),:min(A.shape[0],A.shape[1])] = np.diag(SS)

print('\nU = ')
pprint(UU)
print('\nS = ')
pprint(SSS)
print('\nV = ')
pprint(VVT.T)
print('\nOur SVD is equal to A: ',np.allclose(A, UU @ SSS @ VVT))

print('Check for othogonality of U and V')
print(UU@UU.T)
print(UU.T@UU)
print(VVT@VVT.T)
print(VVT.T@VVT)
````

So now we have an SVD, but this is not made up of the eigenvalues and eigenvectors of $A$. So in the non-symmetric case the SVD is a different decomposition to the eigendecomposition. The fact that $U$ and $V$ are orhtogonal gives us a means to interpret multiplication by $A$ via a rotation and scaling and then a different rotation.
